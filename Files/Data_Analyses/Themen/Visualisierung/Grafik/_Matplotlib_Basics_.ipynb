# Matplotlib Basics

### 1. **Matplotlib importieren**

Zuerst muss die Bibliothek Matplotlib importiert werden, um sie verwenden zu können.
```python
import matplotlib.pyplot as plt
```

`plt` ist dabei die allgemein akzeptierte Abkürzung.

### 2. **Daten vorbereiten**

Als nächstes müssen die Daten bereit gestellt werden, die Analysiert werden sollen. Dies kann über verschiedene Wege erfolgen:

- Der Import von Datensätzen aus einer Datei, Datenbank oder über einen Link.
- Das random erstellen von Daten durch unterschiedliche Funktionen / Methoden.
- etc.

```python
# Beispiel-Daten
x = [1, 2, 3, 4, 5]  # x-Achsenwerte
y = [1, 4, 9, 16, 25]  # y-Achsenwerte
```

Außerdem findet hier das Aufbereiten der Daten statt:

- Löschen oder Ändern von NaN's
- Ändern von Datentypen
- Löschen oder Ändern von Ausreißern
- ...


### 3. **Eine Grafik (Figure) erstellen**

Mit der Methode `.figure()`, (Figure = Grafikfläche) wird die Grundlage des Plot gesetzt.

```python
plt.figure(figsize=(8, 6))  # Größe der Grafik (8x6 Inches)
```

Über die Parameter `figsize` kann man die Größe der Grafik in Inches einstellen.

### 4. **Daten plotten**

Hier plotten wir die Datenpunkte, in diesem Fall als einfache Linie. Dies ist der eigentliche Plot-Befehl.

```python
plt.plot(x, y, label='Quadratische Werte', color='blue')  # Linie mit Farbe und Label
```

### 5. **Titel und Achsenbeschriftungen hinzufügen**

Damit die Grafik verständlicher wird, solltest du einen Titel und Achsenbeschriftungen hinzufügen.

```python
plt.title("Beispiel: Quadratische Funktion")  # Titel der Grafik
plt.xlabel("X-Achse")  # Beschriftung der X-Achse
plt.ylabel("Y-Achse")  # Beschriftung der Y-Achse
```

### 6. **Legende und Gitter hinzufügen** (Optional)

Eine Legende kann hilfreich sein, um die geplotteten Daten besser zu verstehen, und ein Gitter erhöht die Lesbarkeit der Grafik.

```python
plt.legend()  # Zeigt die Legende an
plt.grid(True)  # Fügt ein Gitter hinzu
```

### 7. **Grafik anzeigen**

Zum Schluss zeigst du die erstellte Grafik an.

```python
plt.show()  # Zeigt die Grafik an
```

### Zusammengefasst in einem Code:

````python
import matplotlib.pyplot as plt

# Beispiel-Daten
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Grafik erstellen
plt.figure(figsize=(8, 6))

# Daten plotten
plt.plot(x, y, label='Quadratische Werte', color='blue')

# Titel und Achsenbeschriftungen
plt.title("Beispiel: Quadratische Funktion")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")

# Legende und Gitter
plt.legend()
plt.grid(True)

# Grafik anzeigen
plt.show()
````
