# Piechart (Kuchendiagramm) mit matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei einlesen
df = pd.read_csv('Automobile_data.csv')

# Zählen der Vorkommen von 'fuel-type' (Benzin vs. Diesel)
fuel_counts = df['fuel-type'].value_counts()

# Filtern der Daten auf Benzin und Diesel
fuel_counts = fuel_counts[fuel_counts.index.isin(['gas', 'diesel'])]

# Kuchendiagramm erstellen
plt.figure(figsize=(6, 6))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])

# Titel hinzufügen
plt.title('Verteilung von Benzin- und Dieselfahrzeugen')

# Diagramm anzeigen
plt.axis('equal')  # Damit das Diagramm rund bleibt
plt.show()
```

# Piechart (Kuchendiagramm) mit Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV-Datei einlesen
df = pd.read_csv('Automobile_data.csv')

# Zählen der Vorkommen von 'fuel-type' (Benzin vs. Diesel)
fuel_counts = df['fuel-type'].value_counts()

# Filtern der Daten auf Benzin und Diesel
fuel_counts = fuel_counts[fuel_counts.index.isin(['gas', 'diesel'])]

# Seaborn-Stil anwenden
sns.set(style="whitegrid")

# Kuchendiagramm erstellen mit Seaborn-Stil
plt.figure(figsize=(6, 6))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'orange'])

# Titel hinzufügen
plt.title('Verteilung von Benzin- und Dieselfahrzeugen (Seaborn-Stil)')

# Diagramm anzeigen
plt.axis('equal')  # Damit das Diagramm rund bleibt
plt.show()
```
