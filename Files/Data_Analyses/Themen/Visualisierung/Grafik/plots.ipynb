```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Syllabus

Utilize Matplotlib and Seaborn to create various types of plots, including Boxplots, Histograms, Scatterplots, Lineplots, and Correlation heatmaps.

### 1. Boxplot (Box-and-Whisker Plot)
Ein **Boxplot** zeigt die Verteilung der Daten auf Basis von fünf Kennzahlen: Minimum, erstes Quartil (Q1), Median, drittes Quartil (Q3) und Maximum. Es eignet sich gut, um Ausreißer zu identifizieren.



```python
# Beispiel-Daten
data = np.random.normal(0, 1, 100)

# Boxplot mit Seaborn
plt.figure(figsize=(6, 4))
sns.boxplot(data=data)
plt.title('Boxplot of Normally Distributed Data')
plt.show()
```

In diesem Plot zeigt die Box den Bereich zwischen Q1 und Q3, der Median wird als Linie in der Box dargestellt, und die "Whiskers" (Linien außerhalb der Box) erstrecken sich bis zum Minimum bzw. Maximum (ohne Ausreißer).

### 2. Histogramm
Ein **Histogramm** zeigt die Häufigkeitsverteilung einer einzelnen Variablen. Es ist ideal, um die Verteilung der Daten zu visualisieren.



```python
# Histogramm mit Seaborn
plt.figure(figsize=(6, 4))
sns.histplot(data, bins=20, kde=True)
plt.title('Histogram with KDE of Normally Distributed Data')
plt.show()
```

Das Histogramm zeigt die Häufigkeit der Datenpunkte, die in jede bin (Intervall) fallen. Der KDE (Kernel Density Estimate) überlagert die Dichte der Daten.

### 3. Scatterplot
Ein **Scatterplot** wird verwendet, um die Beziehung zwischen zwei Variablen zu visualisieren. Jeder Punkt repräsentiert ein Datenpaar aus zwei Variablen.



```python
# Beispiel-Daten
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)

# Scatterplot mit Seaborn
plt.figure(figsize=(6, 4))
sns.scatterplot(x=x, y=y)
plt.title('Scatterplot of x vs y')
plt.show()
```

Hier zeigt der Scatterplot die lineare Beziehung zwischen den beiden Variablen `x` und `y`.

### 4. Lineplot
Ein **Lineplot** wird oft verwendet, um Datenpunkte über einen kontinuierlichen Zeitraum oder in Abhängigkeit von einer anderen kontinuierlichen Variable zu visualisieren. Er ist besonders nützlich für Zeitreihendaten.

```python
# Lineplot mit Seaborn
plt.figure(figsize=(6, 4))
sns.lineplot(x=x, y=y)
plt.title('Lineplot of x vs y')
plt.show()
```

In diesem Beispiel verbindet der Lineplot die Datenpunkte durch eine Linie, was besonders nützlich für Zeitreihendaten ist.

### 5. Korrelations-Heatmap
Eine **Heatmap** ist ein guter Weg, um die Korrelationen zwischen mehreren Variablen zu visualisieren. Hierbei werden die Korrelationskoeffizienten als Farben dargestellt.



```python
# Beispiel-Daten
data = np.random.rand(100, 3)
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Korrelationsmatrix
corr = df.corr()

# Heatmap mit Seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

Die Heatmap zeigt, wie stark zwei Variablen miteinander korrelieren. Positive Werte bedeuten eine positive Korrelation, während negative Werte auf eine negative Korrelation hinweisen.

---

### Zusammenfassung
- **Boxplot**: Zeigt die Verteilung und Ausreißer einer einzelnen Variablen.
- **Histogramm**: Zeigt die Häufigkeitsverteilung einer einzelnen Variablen.
- **Scatterplot**: Visualisiert die Beziehung zwischen zwei Variablen.
- **Lineplot**: Zeigt Datenpunkte über eine kontinuierliche Achse, z.B. Zeitreihen.
- **Korrelations-Heatmap**: Zeigt die Korrelationen zwischen mehreren Variablen.
