# KNN

K-Nearest Neighbors (KNN) ist ein einfacher, instanzbasierter Algorithmus, der zur Klassifikation oder Regression verwendet wird, indem er die k n√§chsten Datenpunkte im Merkmalsraum identifiziert und deren Mehrheitsklasse oder Mittelwert als Vorhersage verwendet.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# Pinguin-Datensatz laden
penguins = sns.load_dataset('penguins')

# Entferne Zeilen mit fehlenden Werten
penguins = penguins.dropna()

# Visualisierung der Daten (nur 2 Features f√ºr einfache Darstellung)
sns.pairplot(penguins, hue='species', vars=['bill_length_mm', 'flipper_length_mm'])
plt.show()

# Features und Zielvariable definieren
X = penguins[['bill_length_mm', 'flipper_length_mm']]
y = penguins['species']

# Daten in Trainings- und Testset aufteilen
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# KNN-Klassifikator erstellen und trainieren
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Vorhersagen machen
y_pred = knn.predict(X_test)

# Ergebnisse anzeigen
print(classification_report(y_test, y_pred))
```
