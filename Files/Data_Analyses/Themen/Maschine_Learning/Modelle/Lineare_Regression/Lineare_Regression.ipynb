# Lineare Regression

Die lineare Regression ist eine Methode im maschinellen Lernen und der Statistik, die verwendet wird, um eine Beziehung zwischen zwei oder mehr Variablen zu modellieren. Die Idee dahinter ist einfach: Man versucht, eine Gerade durch die Datenpunkte zu legen, die die Beziehung zwischen einer unabhängigen Variablen (Eingabe) und einer abhängigen Variablen (Ziel) beschreibt.

### Grundidee der Linearen Regression

- **Eingangsvariable (Feature)**: Dies ist die Variable, die zur Vorhersage verwendet wird. Zum Beispiel die Wohnfläche eines Hauses.
- **Zielvariable**: Dies ist die Variable, die vorhergesagt wird. Zum Beispiel der Preis des Hauses.

In der einfachsten Form der linearen Regression wird angenommen, dass die Beziehung zwischen der Eingangs- und der Zielvariable linear ist. Das bedeutet, man nimmt an, dass die Veränderung in der Zielvariable proportional zur Veränderung in der Eingangsvariable ist.

Die lineare Regression versucht, die beste lineare Gleichung zu finden, die diese Beziehung beschreibt:

$$
y = m \cdot x + b
$$

Hier sind:
- $y$ : Die vorhergesagte Zielvariable (z. B. der Preis).
- $x$ : Die Eingangsvariable (z. B. die Wohnfläche).
- $m$ : Die Steigung, die zeigt, wie stark sich $y$ ändert, wenn $x$ sich um eine Einheit verändert.
- $b$ : Der y-Achsenabschnitt der anzeigt, wo auf der y-Achse die Steigung beginnt.

### Ziel der Linearen Regression

Das Ziel der linearen Regression ist es, die Werte von $m$ und $b$ zu finden, die die Summe der quadratischen Abweichungen zwischen den tatsächlichen Werten und den vorhergesagten Werten minimieren. Diese Summe wird durch eine Verlustfunktion bzw.[[Minimierung]] berechnet.

# Praxisbeispiel

Ein einfaches Beispiel für die Anwendung der Linearen Regression wäre die Berechnung des Hauspreises, basierend auf der Wohnfläche:

| Wohnfläche (m²) | Preis (in 1.000 €) |
|-----------------|--------------------|
| 50              | 150                |
| 60              | 180                |
| 70              | 210                |
| 80              | 240                |
| 90              | 270                |

Hier könnte eine lineare Gleichung den Zusammenhang zwischen Wohnfläche und Preis beschreiben. Die lineare Regression würde dann eine gerade Linie durch diese Punkte ziehen, die diesen Zusammenhang am besten beschreibt.

### Python-Code zur linearen Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Daten: Wohnfläche (X) und Preis (y)
X = np.array([[50], [60], [70], [80], [90]])  # Wohnfläche in Quadratmetern
y = np.array([150, 180, 210, 240, 270])       # Preis in 1.000 €

# Modell erstellen
model = LinearRegression()

# Modell auf die Daten trainieren
model.fit(X, y)

# Vorhersagen mit dem Modell
y_pred = model.predict(X)

# Steigung und Achsenabschnitt
print("Steigung (m):", model.coef_[0])
print("Achsenabschnitt (b):", model.intercept_)

# Plot der Daten und der Regressionslinie
plt.scatter(X, y, color='blue', label='Daten')
plt.plot(X, y_pred, color='red', label='Regressionslinie')
plt.scatter(neue_wohnflaeche, vorhergesagter_preis, color='green', label='Vorhersage (75 m²)')
plt.xlabel('Wohnfläche (m²)')
plt.ylabel('Preis (in 1.000 €)')
plt.legend()
plt.show()
```

In diesem Beispiel:
- `model.coef_` gibt die Steigung $m$ der Regressionslinie an.
- `model.intercept_` gibt den Achsenabschnitt $b$ an.

# Aufräumen

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Erstellen von Zufallsdaten für zwei unabhängige Variablen X1 und X2
np.random.seed(42)
n = 100
X1 = np.random.rand(n) * 10  # Zufallswerte für X1
X2 = np.random.rand(n) * 10  # Zufallswerte für X2

# Setzen der Koeffizienten
beta_0 = 2
beta_1 = 1.5
beta_2 = -0.8

# Erstellen des abhängigen Wertes Y mit einem Fehlerterm
epsilon = np.random.randn(n) * 2  # Fehlerterm
Y = beta_0 + beta_1 * X1 + beta_2 * X2 + epsilon

# In ein DataFrame packen für die Visualisierung
data = pd.DataFrame({'X1': X1, 'X2': X2, 'Y': Y})

# Plotten der linearen Regression für X1 und Y
sns.lmplot(x="X1", y="Y", data=data)
plt.title('Lineare Regression: Y gegen X1')
plt.show()

# Plotten der linearen Regression für X2 und Y
sns.lmplot(x="X2", y="Y", data=data)
plt.title('Lineare Regression: Y gegen X2')
plt.show()
```

### Formel:

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_k X_k + \epsilon
$$

$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_k X_k + \epsilon$

### Legende:

$Y$ : Abhängige Variable (das, was vorhergesagt oder erklärt wird).

$\beta_0$ : Achsenabschnitt (Intercept), der konstante Wert, wenn alle $X$-Werte 0 sind.

$\beta_1, \beta_2, \dots, \beta_k$ : Regressionskoeffizienten, die den Einfluss der jeweiligen unabhängigen Variablen $X_1, X_2, \dots, X_k$ auf $Y$ beschreiben.

$X_1, X_2, \dots, X_k$ : Unabhängige Variablen (Prädiktoren), die verwendet werden, um $Y$ vorherzusagen.

$\epsilon$ : Störterm (Fehlerterm), der den Teil von $Y$ beschreibt, der nicht durch die unabhängigen Variablen erklärt wird.

### Zusammenfassung

$\beta_0$ = 2 (Y-Achsenabschnitt, der uns sagt, an welcher Stelle der Y-Achse die Gerade beginnt.) 

$\beta_1$ = 1.5 (Faktor / positive / proportionale Steigung, um die $Y$ ansteigt (1.5) wenn $X_1$ um 1 steigt.)

$\beta_2$ = -0.8 (Faktor / negative / antiproportionale Steigung, um die $Y$ sinkt (-0.8) wenn $X_2$ um 1 steigt.)

$X_1$ und $X_2$ = jeweils 100 Zufallswerte die in der Formel mit den Faktoren multipliziert werden.



Wenn du sagst, dass $X_1 = 5$ ist, dann kannst du $Y$ berechnen, indem du den Wert in die Regressionsgleichung einsetzt. Hier ist, wie das funktioniert:

### Beispiel:

Nehmen wir an, du hast die folgende Gleichung:

$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2$

Angenommen, die Koeffizienten sind:

- $\beta_0 = 2$
- $\beta_1 = 1.5$
- $\beta_2 = -0.8$

### Schritt-für-Schritt-Berechnung:

Wenn du nun $X_1 = 5$ und einen bestimmten Wert für $X_2$ (z.B. $X_2 = 3$) verwendest, kannst du $Y$ berechnen:

$Y = 2 + 1.5 \cdot 5 - 0.8 \cdot 3$

$Y = 7.1$

2. **Interpretation**:
   - Der berechnete Wert $Y$ beträgt 7.1. Das bedeutet, wenn $X_1 = 5$ und $X_2 = 3$, dann würde der geschätzte Wert von $Y$ 7.1 sein.

3. **Grafische Darstellung**:
   - In einem Diagramm kannst du diesen Punkt $(5, 7.1)$ auf der Regressionsgeraden ablesen, und es zeigt, wo die Gerade bei $X_1 = 5$ den Wert von $Y$ annimmt.

### Zusammenfassung:
Wenn du den Wert von $X_1$ änderst, kannst du den entsprechenden $Y$-Wert aus der Gleichung berechnen und damit auf der Regressionsgeraden sehen, wo dieser Punkt liegt.

# Kleines Praxisbeispiel zur Linearen Regression

### Beispiel: Vorhersage von Immobilienpreisen

**Ziel**: Ein Modell zu trainieren, das den Preis eines Hauses auf Basis seiner Wohnfläche (in Quadratmetern) vorhersagen kann.

#### Daten
Stellen wir uns vor, wir haben eine kleine Datensammlung mit echten Hauspreisen und deren Wohnflächen in Quadratmetern. Die Daten könnten folgendermaßen aussehen:

| Wohnfläche (m²) | Preis (in 1.000 €) |
|-----------------|--------------------|
| 50              | 150                |
| 60              | 180                |
| 70              | 210                |
| 80              | 240                |
| 90              | 270                |

Hier sehen wir eine lineare Beziehung: Für jedes zusätzliche Quadratmeter Wohnfläche steigt der Preis um 30.000 €.

#### Ziel des Modells
Unser Ziel ist es, ein Modell zu trainieren, das die Beziehung zwischen Wohnfläche und Preis erkennt, sodass es für eine neue, nicht in den Trainingsdaten enthaltene Wohnfläche den Preis vorhersagen kann.

### Beispielcode zur Implementierung in Python

Wir verwenden `scikit-learn`, um eine einfache lineare Regression auf diese Daten anzuwenden.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Daten: Wohnfläche (X) und Preis (y)
X = np.array([[50], [60], [70], [80], [90]])  # Wohnfläche in Quadratmetern
y = np.array([150, 180, 210, 240, 270])       # Preis in 1.000 €

# Modell erstellen
model = LinearRegression()

# Modell auf die Daten trainieren
model.fit(X, y)

# Modellvorhersagen für die Trainingsdaten
y_pred = model.predict(X)

# Neue Vorhersage: Preis für ein Haus mit 75 m² Wohnfläche
neue_wohnflaeche = np.array([[75]])
vorhergesagter_preis = model.predict(neue_wohnflaeche)

print("Vorhergesagter Preis für ein Haus mit 75 m² Wohnfläche:", vorhergesagter_preis[0], "Tausend Euro")

# Plot der Daten und der Regressionslinie
plt.scatter(X, y, color='blue', label='Daten')
plt.plot(X, y_pred, color='red', label='Regressionslinie')
plt.scatter(neue_wohnflaeche, vorhergesagter_preis, color='green', label='Vorhersage (75 m²)')
plt.xlabel('Wohnfläche (m²)')
plt.ylabel('Preis (in 1.000 €)')
plt.legend()
plt.show()
```

### Erklärung des Codes

- Wir erstellen die Eingabedaten `X` (Wohnfläche) und die Zielwerte `y` (Preis).
- Ein lineares Regressionsmodell wird mit `LinearRegression()` erstellt.
- Mit `model.fit(X, y)` wird das Modell auf die vorhandenen Daten trainiert.
- Danach nutzen wir `model.predict()`, um den Preis für eine Wohnfläche von 75 m² vorherzusagen.

### Ergebnis
Das Modell sollte in der Lage sein, den Preis eines Hauses mit 75 m² Wohnfläche korrekt oder zumindest annähernd vorherzusagen, basierend auf dem linearen Trend, den es während des Trainings erkannt hat.
