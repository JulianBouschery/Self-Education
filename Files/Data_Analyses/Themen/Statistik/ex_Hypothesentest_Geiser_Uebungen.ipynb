# Hypothesentest und Fehlerarten

## 1. Beschreibe einen Hypothesentest

Ein Hypothesentest ist ein statistisches Verfahren, das verwendet wird, um zu entscheiden, ob eine bestimmte Hypothese über eine Grundgesamtheit auf Basis von Stichprobendaten abgelehnt werden sollte oder nicht.

- **Nullhypothese (H₀)**: Dies ist die Annahme, die getestet wird, und üblicherweise die Aussage darstellt, dass es keinen Effekt oder Unterschied gibt. Beispiel: „Der Mittelwert der Grundgesamtheit ist gleich 24.“
- **Alternativhypothese (H₁)**: Diese Hypothese stellt das Gegenteil der Nullhypothese dar. Beispiel: „Der Mittelwert der Grundgesamtheit ist nicht gleich 24.“

Der Test berechnet einen **Teststatistikwert** (z.B. z- oder t-Wert), der dann verwendet wird, um den **p-Wert** zu bestimmen. Der p-Wert gibt die Wahrscheinlichkeit an, dass die beobachteten Daten unter der Annahme der Nullhypothese zufällig auftreten könnten. Wenn der p-Wert kleiner als das festgelegte Signifikanzniveau (meistens 5%, also \(\alpha = 0.05\)) ist, wird die Nullhypothese abgelehnt.

## 2. Was ist der Fehler 1. Art, was ist der Fehler 2. Art?

- **Fehler 1. Art (Alpha-Fehler)**: Dies ist der Fehler, der auftritt, wenn die Nullhypothese abgelehnt wird, obwohl sie in Wirklichkeit wahr ist. Mit anderen Worten, man nimmt an, dass es einen signifikanten Unterschied oder Effekt gibt, obwohl dieser nicht existiert. Die Wahrscheinlichkeit für diesen Fehler wird durch das Signifikanzniveau $(\alpha\)$ angegeben (z.B. 5%).

- **Fehler 2. Art (Beta-Fehler)**: Dies ist der Fehler, der auftritt, wenn die Nullhypothese beibehalten wird, obwohl sie in Wirklichkeit falsch ist. Das bedeutet, man übersieht einen tatsächlichen Effekt oder Unterschied. Die Wahrscheinlichkeit, einen Fehler 2. Art zu begehen, wird durch \(\beta\) dargestellt. Die Teststärke (Power) eines Tests ist \(1 - \beta\) und gibt an, wie wahrscheinlich es ist, einen tatsächlichen Effekt zu entdecken.

## 3. Test zur Signifikanz des Mittelwerts einer Stichprobe von 100 Werten

Angenommen, du hast eine Stichprobe mit 100 Werten, einem Mittelwert \(\bar{x}\) und einer Standardabweichung \(s\), und du sollst die Signifikanz des Mittelwerts mit einem Signifikanzniveau von 5% testen.

Wenn die Stichprobe groß genug ist (\(n > 30\)), wird üblicherweise der **z-Test** verwendet. Der z-Test setzt voraus, dass entweder die Varianz der Grundgesamtheit bekannt ist oder dass die Stichprobe groß genug ist, um die Normalverteilung anzunehmen.

### Vorgehensweise für einen z-Test:
- Stichprobengröße \(n = 100\)
- Stichprobenmittelwert \(\bar{x}\)
- Standardabweichung der Stichprobe \(s\)
- Mittelwert der Grundgesamtheit \(\mu_0\)
- Signifikanzniveau \(\alpha = 0.05\)

Die Z-Teststatistik wird berechnet mit der Formel:

\[
Z = \frac{\bar{x} - \mu_0}{\frac{\sigma}{\sqrt{n}}}
\]

Falls die Varianz der Grundgesamtheit nicht bekannt ist und die Stichprobe weniger als 30 Beobachtungen enthält, sollte ein **t-Test** verwendet werden, da dieser die Stichprobenstandardabweichung \(s\) verwendet.

In diesem Fall, bei einer Stichprobe von 100 Werten und einem Signifikanzniveau von 5%, wäre der **z-Test** angemessen.


## 4. Python-Programm für den Test zur Signifikanz des Mittelwerts einer Stichprobe von 100 Werten


```python
import numpy as np
from scipy import stats

# Beispieldaten: größere Stichprobe (hier simuliert mit Zufallsdaten)
# Simuliert eine Stichprobe von 100 Werten mit einem Mittelwert von 25 und einer Standardabweichung von 3
np.random.seed(0)  # Für reproduzierbare Ergebnisse
stichprobe = np.random.normal(loc=25, scale=3, size=100)

# Berechnung des Mittelwerts der Stichprobe
mittelwert = np.mean(stichprobe)
standardabweichung = np.std(stichprobe, ddof=1)  # Standardabweichung der Stichprobe

print("Der Mittelwert der Stichprobe ist:", mittelwert)

# Grundgesamtheit: Mittelwert 24 (angenommen)
popmean = 24
pop_std = 3  # Angenommene Standardabweichung der Grundgesamtheit

# Z-Test durchführen
z_stat = (mittelwert - popmean) / (pop_std / np.sqrt(len(stichprobe)))
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_stat)))  # Zweiseitiger Test

print(f'Z-Statistik: {z_stat}, p-Wert: {p_value}')

# Entscheidung auf Basis des Signifikanzniveaus alpha
alpha = 0.05
if p_value < alpha:
    print('Die Nullhypothese wird abgelehnt (es gibt einen signifikanten Unterschied).')
else:
    print('Die Nullhypothese wird beibehalten (kein signifikanter Unterschied).')
```

### Szenario:
Angenommen, wir führen einen Hypothesentest durch, um zu entscheiden, ob eine Maschine richtig kalibriert ist oder nicht.

- **Nullhypothese $ H_0 $**: Die Maschine ist korrekt kalibriert (z.B., der Mittelwert der produzierten Teile beträgt 100).
- **Alternativhypothese $ H_1 $**: Die Maschine ist falsch kalibriert (z.B., der Mittelwert ist nicht 100).

### Fehler 1. Art (Alpha-Fehler):
Fehler 1. Art tritt auf, wenn wir die Nullhypothese ablehnen (d.h., wir denken, die Maschine sei falsch kalibriert), obwohl sie tatsächlich korrekt kalibriert ist. Das Risiko dieses Fehlers ist das Signifikanzniveau $ \alpha $.

### Fehler 2. Art (Beta-Fehler):
Fehler 2. Art tritt auf, wenn wir die Nullhypothese beibehalten (d.h., wir denken, die Maschine sei korrekt kalibriert), obwohl sie tatsächlich falsch kalibriert ist. Das Risiko dieses Fehlers ist $ \beta $.

### Beispiel mit Kurve:
- Die Nullhypothese $ H_0 $ geht davon aus, dass der Mittelwert $ \mu_0 = 100 $ ist.
- Die Alternativhypothese $ H_1 $ nimmt an, dass der Mittelwert z.B. $ \mu_1 = 105 $ ist.
- Der **Alpha-Fehler-Bereich** (Fehler 1. Art) liegt in dem Bereich, in dem wir $ H_0 $ ablehnen, obwohl $ \mu_0 $ gilt.
- Der **Beta-Fehler-Bereich** (Fehler 2. Art) liegt in dem Bereich, in dem wir $ H_0 $ beibehalten, obwohl $ \mu_1 $ gilt.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameter für die Nullhypothese und die Alternativhypothese
mu0 = 100  # Mittelwert unter H0
mu1 = 105  # Mittelwert unter H1
sigma = 5  # Standardabweichung
alpha = 0.05  # Signifikanzniveau

# Z-Wert für das Signifikanzniveau
z_alpha = stats.norm.ppf(1 - alpha)

# Bereich der x-Achse (Teststatistik)
x = np.linspace(85, 120, 500)

# Verteilungen unter H0 und H1
pdf_h0 = stats.norm.pdf(x, mu0, sigma)
pdf_h1 = stats.norm.pdf(x, mu1, sigma)

# Plot
plt.figure(figsize=(10, 6))

# H0-Verteilung
plt.plot(x, pdf_h0, label="Verteilung unter H0 (mu=100)", color="blue")

# H1-Verteilung
plt.plot(x, pdf_h1, label="Verteilung unter H1 (mu=105)", color="red")

# Alpha-Bereich (Fehler 1. Art)
plt.fill_between(x, 0, pdf_h0, where=(x > mu0 + z_alpha * sigma), color='blue', alpha=0.3, label="Alpha-Bereich (Fehler 1. Art)")

# Beta-Bereich (Fehler 2. Art)
plt.fill_between(x, 0, pdf_h1, where=(x < mu0 + z_alpha * sigma), color='red', alpha=0.3, label="Beta-Bereich (Fehler 2. Art)")

# Linien und Beschriftungen
plt.axvline(mu0 + z_alpha * sigma, color='black', linestyle='--', label=f'Grenze bei z = {z_alpha:.2f}')
plt.title('Fehler 1. und 2. Art in einem Hypothesentest')
plt.xlabel('Teststatistik')
plt.ylabel('Dichte')
plt.legend()
plt.grid(True)
plt.show()
```
