# z-Verteilung und t-Verzeilung

- sind jeweils zwei wichtige statistische Verteilungen
- werden für Hypothesentests und Konfidenzintervalle verwendet
- beide Verteilungen helfen bei der Abschätzung von Unsicherheiten
- unterscheiden sich aber in ihren Anwendungsbereichen und Eigenschaften
- t-Verteilung geht bei kleineren Stichproben aufgrund ihrer breiteren Form vorsichtiger mit extremen Werten um

<br>

## z-Verteilung (= Standardnormalverteilung)

- **z-Verteilung** ist eine andere Bezeichnung für **Standardnormalverteilung**
- ist eine spezielle Form der Normalverteilung mit einem Mittelwert von 0 und einer Standardabweichung von 1
- wird verwendet, wenn:
  - die Stichprobenanzahl groß ist (meistens n > 30)
  - die Standardabweichung der Grundgesamtheit bekannt ist.

<br>


## t-Verteilung (= Student's-Verteilung)

- **t-Verteilung** ist eine andere Bezeichnung für 'Student's t-Verteilung'
- ähnelt der Normalverteilung, ist aber bei kleinen Stichproben (meistens n ≤ 30) und unbekannter Standardabweichung der Grundgesamtheit besser geeignet
- sie ist breiter und hat dickere Ränder (dickere 'Schwänze') als die Normalverteilung
  - bedeutet: größere Wahrscheinlichkeiten für extreme Werte
  - das wird der Unsicherheit bei kleineren Stichproben gerecht
- mit zunehmender Stichprobengröße (n > 30), wird sie immer symmetrischer und ähnelt dann der z-Verteilung
- sie ist abhängig von den sogenannten **Freiheitsgraden** (degrees of freedom, df), die in der Regel n - 1 sind

<br>

Beispiel:
- man hat eine kleine Stichprobe (n = 10) mit einem Mittelwert von 100 und einer Standardabweichung von 15
- die wahre Standardabweichung der Grundgesamtheit ist unbekannt 
- man benutzt die t-Verteilung, um das Konfidenzintervall oder den t-Wert zu berechnen
- der t-Wert hängt von den Freiheitsgraden (n - 1) und dem gewünschten Konfidenzniveau ab

<br>

## Zusammenfassung der Unterschiede

| Eigenschaft          | Z-Verteilung                   | T-Verteilung                         |
|----------------------|--------------------------------|--------------------------------------|
| **Anwendung**        | Große Stichproben, bekannte Standardabweichung | Kleine Stichproben, unbekannte Standardabweichung |
| **Form**             | Schmalere Verteilung           | Breitere Verteilung mit dickeren Rändern |
| **Freiheitsgrade**   | Nicht benötigt                 | n - 1 Freiheitsgrade                 |
| **Annäherung**       | Ähnelt der t-Verteilung bei großen Stichproben | Wird der z-Verteilung bei großen Stichproben ähnlich |

# Konfidenzintervall

- auch Vertrauensintervall (Confidence Interval, CI)
- dient zur Beurteilung der Datenzuverlässigkeit
- ist ein Bereich von Werten, der mit einer bestimmten Wahrscheinlichkeit den tatsächlichen Parameter (z.B. Mittelwert) einer Population einschließt
- typische Vertrauensintervalle sind das **95%-** oder **99%-Konfidenzintervall**
   - bedeutet: wir sind zu 95% bzw. 99% sicher, dass der tatsächliche Parameter innerhalb dieses Bereichs liegt

<br>
 
Das Vertrauensintervall für den Mittelwert einer Normalverteilung ist gegeben durch:

<div style="margin-left: 40px;">
$ CI = \bar{x} \pm z \frac{\sigma}{\sqrt{n}} $
</div>

- $ \bar{x} $ = Stichprobenmittelwert
- $ z $ = [z-Wert](PCED/Block_3/Grundbegriffe.ipynb#Z_Verteilung) (abhängig vom Vertrauensniveau, z.B. 1.96 für 95%)
- $ \sigma $ = (Sigma) Standardabweichung der Population (falls bekannt) oder der Stichprobe
- $ n $ = Stichprobengröße

<br>

- **wenn n < 30:**
  - t-Verteilung (= 'Student-Verteilung')
  - diese berücksichtigt die Unsicherheit bei kleineren Stichproben und unbekannter Populationsstandardabweichung<br><br>
- **wenn n ≥ 30 und Standardabweichung der Population bekannt:**
  - z-Verteilung (Normalverteilung) kann verwendet werden
  - da größere Stichproben (n ≥ 30) dazu neigen, der Normalverteilung zu folgen (per Zentraler Grenzwertsatz)<br><br>
- **wenn n ≥ 30 und Standardabweichung der Population nicht bekannt:**
  - Standardabweichung der Stichprobe kann als Näherung verwendet werden
  - aber für kleinere Stichproben sollte trotzdem die t-Verteilung verwendet werden

## Beispiel 1: Konfidenzintervalle mit z-Verteilung

### Beispiel Gewicht
- wir haben eine Stichprobe von n = 100 Personen
- das Durchschnittsgewicht beträgt 75 kg
- eine bekannte Standardabweichung der Population von 10 kg
- wir wollen ein 95 %-Konfidenzintervall für das durchschnittliche Gewicht der Population berechnen

````python
import math
from scipy.stats import norm

# Daten für das z-Konfidenzintervall
n_z = 100  # Stichprobengröße
mean_z = 75  # Mittelwert der Stichprobe
sigma_z = 10  # Standardabweichung der Population
confidence_level_z = 0.95  # 95% Konfidenzintervall

# z-Wert für das 95% Konfidenzniveau
z_value = norm.ppf(1 - (1 - confidence_level_z) / 2)

# Konfidenzintervallberechnung für z-Verteilung
margin_of_error_z = z_value * (sigma_z / math.sqrt(n_z))
lower_bound_z = mean_z - margin_of_error_z
upper_bound_z = mean_z + margin_of_error_z

print(f"z-Konfidenzintervall: [{lower_bound_z:.2f}, {upper_bound_z:.2f}]")

# Werte für den Plot
x = np.linspace(50, 90, 400)  # Bereich der x-Werte für den Plot
y = stats.norm.pdf(x, mean_z, sigma_z)  # Normalverteilung

# Plot erstellen
plt.plot(x, y, label='Normalverteilung', color='blue')
plt.fill_between(x, y, where=(x >= lower_bound_z) & (x <= upper_bound_z), color='lightblue', alpha=0.5, label='95% Konfidenzintervall')
plt.axvline(x=lower_bound_z, color='red', linestyle='--', label='Unteres Limit')
plt.axvline(x=upper_bound_z, color='green', linestyle='--', label='Oberes Limit')
plt.axvline(x=mean_z, color='black', linestyle='-', label='Stichprobenmittelwert')

# Plot-Details
plt.title('Konfidenzintervall für Körpergewicht')
plt.xlabel('Körpergewicht (kg)')
plt.ylabel('Dichte')
plt.legend()
plt.grid()
plt.xlim(50, 90)
plt.ylim(0, 0.05)

plt.show()
````

### Beispiel Simulation

- wir simulieren eine Stichprobe (n = 50) von Daten um dann das 95%-Vertrauensintervall des Mittelwerts zu berechnen
- Mittelwert = 100
- Standardabweichung = 15

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simuliere eine Stichprobe
np.random.seed(0) # Setzt den Zufallsgenerator auf einen festen Startwert (für Reproduzierbarkeit)
stichprobe = np.random.normal(loc=100, scale=15, size=50)  # Mittelwert = 100, Standardabweichung = 15, Stichprobengröße = 50

# Berechne den Stichprobenmittelwert und die Stichprobenstandardabweichung
mittelwert = np.mean(stichprobe)
std_abw = np.std(stichprobe, ddof=1)  # ddof=1 (Degrees of Freedom), um die Stichproben-Standardabweichung zu berechnen, nicht die Populationsstandardabweichung
n = len(stichprobe)

# Berechne das 95%-Vertrauensintervall
z = 1.96  # z-Wert für 95%-Vertrauensniveau
margin_of_error = z * (std_abw / np.sqrt(n)) # Fehlerspielraum gib an, wie weit der wahre Mittelwert der Population vom Mittelwert der Stichprobe abweichen könnte
ci_low = mittelwert - margin_of_error
ci_high = mittelwert + margin_of_error

# Ausgabe des Vertrauensintervalls
print(f"95%-Vertrauensintervall: [{ci_low:.2f}, {ci_high:.2f}]")
print(f"Mittelwert der Stichprobe: {mittelwert:.2f}")

# Werte für den Plot
x = np.linspace(70, 130, 400)
y = stats.norm.pdf(x, mittelwert, std_abw)

# Plot erstellen
plt.plot(x, y, label='Normalverteilung', color='blue')
plt.fill_between(x, y, where=(x >= ci_low) & (x <= ci_high), color='lightblue', alpha=0.5, label='95% Konfidenzintervall')
plt.axvline(x=ci_low, color='red', linestyle='--', label='Unteres Limit')
plt.axvline(x=ci_high, color='green', linestyle='--', label='Oberes Limit')
plt.axvline(x=mittelwert, color='black', linestyle='-', label='Stichprobenmittelwert')

# Plot-Details
plt.title('Konfidenzintervall mit z-Verteilung')
plt.xlabel('Wert')
plt.ylabel('Dichte')
plt.legend()
plt.grid()
plt.xlim(70, 130)
plt.ylim(0, 0.05)

plt.show()
````

- der Mittelwert der Stichprobe und die Stichprobenstandardabweichung werden berechnet
- ein 95%-Konfidenzintervall um den Mittelwert wird berechnet
- dabei wird der z-Wert (1.96) für ein 95%-Konfidenzniveau verwendet
- das Konfidenzintervall gibt den Bereich an, in dem mit 95 %iger Wahrscheinlichkeit der wahre Mittelwert der Population liegt

## Beispiel 2: Konfidenzintervall mit t-Verteilung

### Beispiel Testnoten
- wir haben eine kleinere von n = 15 Studierenden
- deren Durchschnittsnote in einem Test beträgt 80 Punkte
- die Populationsstandardabweichung ist unbekannt
- die Standardabweichung der Stichprobe beträgt 8 Punkte
- wir wollen ein 95 %-Konfidenzintervall für die durchschnittliche Testnote der gesamten Population berechnen

````python
from scipy.stats import t

# Daten für das t-Konfidenzintervall
n_t = 15  # Stichprobengröße
mean_t = 80  # Mittelwert der Stichprobe
s_t = 8  # Standardabweichung der Stichprobe
confidence_level_t = 0.95  # 95% Konfidenzintervall
df_t = n_t - 1  # Freiheitsgrade

# t-Wert für das 95% Konfidenzniveau
t_value = t.ppf(1 - (1 - confidence_level_t) / 2, df_t)

# Konfidenzintervallberechnung für t-Verteilung
margin_of_error_t = t_value * (s_t / math.sqrt(n_t))
lower_bound_t = mean_t - margin_of_error_t
upper_bound_t = mean_t + margin_of_error_t

print(f"t-Konfidenzintervall: [{lower_bound_t:.2f}, {upper_bound_t:.2f}]")

# Plot erstellen
plt.plot(x, y, label='t-Verteilung', color='blue')
plt.fill_between(x, y, where=(x >= lower_bound_t) & (x <= upper_bound_t), color='lightblue', alpha=0.5, label='95% Konfidenzintervall')
plt.axvline(x=lower_bound_t, color='red', linestyle='--', label='Unteres Limit')
plt.axvline(x=upper_bound_t, color='green', linestyle='--', label='Oberes Limit')
plt.axvline(x=mean_t, color='black', linestyle='-', label='Stichprobenmittelwert')

# Plot-Details
plt.title('Konfidenzintervall für Testnoten (t-Verteilung)')
plt.xlabel('Testnoten')
plt.ylabel('Dichte')
plt.legend()
plt.grid()
plt.xlim(50, 100)
plt.ylim(0, 0.05)

plt.show()
````

### Beispiel Simulation

- wir simulieren eine Stichprobe (n = 10)
- aufgrund der kleinen Stichprobe verwenden wir die t-Verteilung, um den Unsicherheitsfaktor zu berücksichtigen.
- wir wollen das 95%-Vertrauensintervall des Mittelwerts berechnen

````python
# Hier habe ich einen Wert von n=10, d.h. Student-Verteilung

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simuliere eine Stichprobe (Ändere die Stichprobengröße auf n=10)
np.random.seed(0)
stichprobe = np.random.normal(loc=100, scale=15, size=10)  # Mittelwert = 100, Standardabweichung = 15, Stichprobengröße = 10

# Berechne den Stichprobenmittelwert und die Stichprobenstandardabweichung
mittelwert = np.mean(stichprobe)
std_abw = np.std(stichprobe, ddof=1)  # ddof=1 für Stichproben-Standardabweichung
n = len(stichprobe)

# Berechne den t-Wert für das 95%-Vertrauensniveau und n-1 Freiheitsgrade
t_wert = stats.t.ppf(0.975, df=n-1)  # 0.975 für zweiseitiges 95%-Intervall

# Berechne den Fehlerbereich (margin of error)
margin_of_error = t_wert * (std_abw / np.sqrt(n))

# Berechne das 95%-Vertrauensintervall
ci_low = mittelwert - margin_of_error
ci_high = mittelwert + margin_of_error

# Ausgabe des Vertrauensintervalls
print(f"95%-Vertrauensintervall: [{ci_low:.2f}, {ci_high:.2f}]")
print(f"Mittelwert der Stichprobe: {mittelwert:.2f}")

# Werte für den Plot
x = np.linspace(60, 140, 200)  # Bereich der x-Werte für den Plot
y = stats.t.pdf((x - mittelwert) / (std_abw / np.sqrt(n)), df=n-1) / (std_abw / np.sqrt(n))  # t-Verteilung

# Plot erstellen
plt.plot(x, y, label='t-Verteilung', color='blue')
plt.fill_between(x, y, where=(x >= ci_low) & (x <= ci_high), color='lightblue', alpha=0.5, label='95%-Konfidenzintervall')
plt.axvline(x=ci_low, color='red', linestyle='--', label='Unteres Limit')
plt.axvline(x=ci_high, color='green', linestyle='--', label='Oberes Limit')
plt.axvline(x=mittelwert, color='black', linestyle='-', label='Stichprobenmittelwert')

# Plot-Details
plt.title('Konfidenzintervall für die Stichprobe (t-Verteilung)')
plt.xlabel('Wert')
plt.ylabel('Dichte')
plt.legend()
plt.grid()
plt.xlim(60, 140)
plt.ylim(0, 0.1)

plt.show()
````

- der Mittelwert der Stichprobe beträgt 111.07
- das 95%-Konfidenzintervall gibt an, dass der wahre Populationsmittelwert mit einer Wahrscheinlichkeit von 95 % zwischen 100.13 und 122.01 liegt
- da die Stichprobengröße klein ist (n = 10), wird die t-Verteilung verwendet
- diese erzeugt breitere Konfidenzintervalle als die z-Verteilung, um die Unsicherheit bei kleinen Stichproben zu berücksichtigen

## Beispiel Gegenüberstellung von t- und z-Verteilung

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simuliere eine Stichprobe mit n < 30 (z.B. n=10)
np.random.seed(0)
stichprobe = np.random.normal(loc=100, scale=15, size=10)  # Mittelwert = 100, Standardabweichung = 15, Stichprobengröße = 10

# Berechne den Stichprobenmittelwert und die Stichprobenstandardabweichung
mittelwert = np.mean(stichprobe)
std_abw = np.std(stichprobe, ddof=1)  # ddof=1 für Stichproben-Standardabweichung
n = len(stichprobe)

# Berechne den t-Wert für das 95%-Vertrauensniveau und n-1 Freiheitsgrade
t_wert = stats.t.ppf(0.975, df=n-1)  # 0.975 für zweiseitiges 95%-Intervall

# Berechne den Fehlerbereich (margin of error) für die t-Verteilung
margin_of_error_t = t_wert * (std_abw / np.sqrt(n))

# Berechne das 95%-Vertrauensintervall für die t-Verteilung
ci_low_t = mittelwert - margin_of_error_t
ci_high_t = mittelwert + margin_of_error_t

# Berechne den Z-Wert (Normalverteilung) für das 95%-Vertrauensniveau
z_wert = 1.96  # Z-Wert für 95% bei Standardnormalverteilung

# Berechne den Fehlerbereich (margin of error) für die Z-Verteilung
margin_of_error_z = z_wert * (std_abw / np.sqrt(n))

# Berechne das 95%-Vertrauensintervall für die Z-Verteilung
ci_low_z = mittelwert - margin_of_error_z
ci_high_z = mittelwert + margin_of_error_z

# Ausgabe des Vertrauensintervalls
print(f"95%-Vertrauensintervall (t-Verteilung): [{ci_low_t:.2f}, {ci_high_t:.2f}]")
print(f"95%-Vertrauensintervall (Z-Verteilung): [{ci_low_z:.2f}, {ci_high_z:.2f}]")
print(f"Mittelwert der Stichprobe: {mittelwert:.2f}")

# Plot der Stichprobenverteilung und der Vertrauensintervalle
plt.figure(figsize=(8, 6))

# Erstellen der Normalverteilung mit dem Mittelwert und der Standardabweichung der Stichprobe
x = np.linspace(mittelwert - 4*std_abw, mittelwert + 4*std_abw, 1000)
y = stats.norm.pdf(x, loc=mittelwert, scale=std_abw)

# Plot der Normalverteilung
plt.plot(x, y, label='Stichprobenverteilung', color='blue')

# Markiere den Bereich des 95%-Vertrauensintervalls (t-Verteilung)
plt.axvline(ci_low_t, color='red', linestyle='--', label=f'Unteres Intervall t-Verteilung ({ci_low_t:.2f})')
plt.axvline(ci_high_t, color='red', linestyle='--', label=f'Oberes Intervall t-Verteilung ({ci_high_t:.2f})')

# Markiere den Bereich des 95%-Vertrauensintervalls (Z-Verteilung)
plt.axvline(ci_low_z, color='green', linestyle='--', label=f'Unteres Intervall Z-Verteilung ({ci_low_z:.2f})')
plt.axvline(ci_high_z, color='green', linestyle='--', label=f'Oberes Intervall Z-Verteilung ({ci_high_z:.2f})')

# Grafikeinstellungen
plt.title(f'95%-Vertrauensintervall: t-Verteilung vs. Z-Verteilung (n = {n})')
plt.xlabel('Werte')
plt.ylabel('Wahrscheinlichkeit')
plt.legend()
plt.grid(True)

# Zeige den Plot
plt.show()
````

# Zentraler Grenzwertsatzes (ZGS)

siehe [hier](PCED/Block_3/Grundbegriffe.ipynb#zentraler_Grundwertsatz)

- besagt, dass die Verteilung der Mittelwerte von genügend großen Stichproben einer Zufallsvariable annähernd normalverteilt ist
- unabhängig von der Form der ursprünglichen Verteilung der Population

# Erforderliche Stichprobengröße ermitteln

- wir berechnen die benötigte Stichprobengröße anhand des zentralen Grenzwertsatzes (ZGS)

````python
# Berechnung der Stichprobengröße mit dem zentralen Grenzwertsatz:
# Hier die Formel:

# https://questionstar.de/blog/berechnung-der-stichprobengrose-technische-details-einfach-erklaert/

import math

# Parameter
N = 1000000  # Grundgesamtheit
Z = 1.96     # Z-Wert für 95%-Konfidenzniveau
p = 0.5      # Konservativ geschätzte Wahrscheinlichkeit
E = 0.05     # Stichprobenfehler (5%)

# Berechnung der Stichprobengröße
numerator = (Z**2) * p * (1 - p) * N # Numerator = Zähler
denominator = (E**2) * (N - 1) + (Z**2) * p * (1 - p) # Denominator = Nenner
n = numerator / denominator

# Ausgabe
print(f"Erforderliche Stichprobengröße: {math.ceil(n)}")
````


Parameter:
- **N**: Grundgesamtheit (die Gesamtzahl der Individuen in der Population, aus der die Stichprobe gezogen wird). In Ihrem Beispiel beträgt \(N = 1{,}000{,}000\).
- **Z**: Z-Wert (z. B. 1.96 für ein 95%-Konfidenzniveau). Dieser Wert repräsentiert die Anzahl der Standardabweichungen, die erforderlich sind, um das gewünschte Konfidenzniveau zu erreichen.
- **p**: Geschätzte Wahrscheinlichkeit (in Ihrem Fall 0.5, was die konservativste Schätzung ist, da sie die größte benötigte Stichprobe ergibt). Dies kann der Anteil einer bestimmten Merkmalsausprägung in der Population sein.
- **E**: Stichprobenfehler (z. B. 0.05, was 5 % entspricht). Dies gibt an, wie viel Abweichung vom wahren Wert der Parameter zulässig ist.

# Hypothesentest

[Hypothesentest, Youtube](https://studyflix.de/statistik/hypothesentest-1719)<br>
[Hypothesentest, Statistik, Wikipedia](https://de.wikipedia.org/wiki/Hypothese_(Statistik))

<br>

**Der Ablauf eines Hypothesentests:**

- Schritt 1: Null- und Alternativhypothese aufstellen
  - wir definieren die Hypothesen basierend auf der Fragestellung

<br>

- Schritt 2: Teststatistik berechnen
  - Teststatistik wird berechnet, um den Unterschied zwischen den beobachteten Daten und der Nullhypothese zu messen

<br>

- Schritt 3: p-Wert bestimmen
  - gibt an, wie wahrscheinlich es ist, dass die beobachteten Daten unter der Annahme der Nullhypothese auftreten

<br>

- Schritt 4: Ergebnis interpretieren
  - wenn p-Wert < vordefiniertes Signifikanzniveau (z.B. 0,05), dann wird die Nullhypothese abgelehnt


## z-Test und t-Test

wir unterscheiden zwei Arten von Hypothesentests:
- z-Test
- t-Test

Beides sind Hypothesentests, haben aber unterschiedliche Voraussetzungen und werden in unterschiedlichen Situationen verwendet.

<br>

### Verwendung
**z-Test:**
- wird verwendet, wenn die Standardabweichung der Population bekannt und die Stichprobengröße ausreichend groß ist
- typisch für große Stichproben $(n≥30)$


**t-Test:**
- wird verwendet, wenn die Standardabweichung der Population nicht bekannt und stattdessen die Stichprobenstandardabweichung geschätzt werden muss
- besonders geeignet für kleine Stichproben $(n<30)$

<br>

### einfache Beispiele
z-Test:<br>
Man führt eine Umfrage unter 1000 Personen durch und weiß, dass die Standardabweichung der Antwortzeit in der gesamten Population 5 Sekunden beträgt. Man möchte testen, ob der Durchschnitt der Antwortzeiten in deiner Stichprobe signifikant von einem theoretischen Durchschnitt abweicht.

t-Test:<br>
Man hat eine Stichprobe von 15 Studierenden und kennt die Standardabweichung der Gesamtpopulation nicht. Man möchte testen, ob der Durchschnitt ihrer Prüfungsergebnisse signifikant von einem festgelegten theoretischen Mittelwert (z.B. 75 Punkte) abweicht.

<br>

## Schritt 1: Nullhypothese und Alternativhypothese

- der Hypothesentest beginnt mit zwei sich gegenseitig ausschließenden Hypothesen:
  - Nullhypothese (H0)
  - Alternativhypothese (H1)

<br>

**Nullhypothese (H0)**:
- ist die Annahme, die getestet werden soll
- sie stellt oft eine Aussage über den Status quo dar
- z.B.: "Der Mittelwert in der Population ist 50".

<br>

**Alternativhypothese (H1)**:
- ist die Aussage, die überprüft werden soll, wenn die Nullhypothese nicht zutrifft
- z.B.: "Der Mittelwert ist nicht 50" (zweiseitig) oder "Der Mittelwert ist größer als 50" (einseitig).

<br>

im Hypothesentest wird versucht zu bestimmen:
- **hat man genügend Beweise, um die Nullhypothese zu verwerfen und die Alternativhypothese zu akzeptieren?**


## Schritt 2: Teststatistik

- wird berechnet, um zu entscheiden, ob die Nullhypothese abgelehnt werden sollte oder nicht
- basierend auf dem Vergleich mit einem kritischen Wert oder dem p-Wert
- sie gibt an, wie weit die Stichprobendaten von der Nullhypothese abweichen
- ist im Grunde das zentrale Element, um einen Hypothesentest durchzuführen und auf Basis der Daten eine fundierte Entscheidung zu treffen
- die Teststatistik folgt einer bestimmten Wahrscheinlichkeitsverteilung, je nachdem, welcher Test durchgeführt wird (z.B. Normalverteilung (= z-Verteilung), t-Verteilung etc.)

## Schritt 3: p-Wert

- ist die Wahrscheinlichkeit, dass die beobachteten Daten auftreten, wenn die Nullhypothese wahr ist
- wenn p-Wert < Signifikanzniveau, wird die Nullhypothese abgelehnt
- Signifikanzniveau (α): ein vorab festgelegter Wert (oft 5 % oder 0,05), der die Wahrscheinlichkeit angibt, die Nullhypothese fälschlicherweise abzulehnen


## Schritt 4: Interpretation

- Hypothese testen:
  - Vergleich des p-Werts mit dem Signifikanzniveau (α)
  - Entscheidung treffen, ob die Nullhypothese abgelehnt werden soll oder nicht

<br>

## Fehlerarten

Bei einem Hypothesentest können zwei Arten von Fehlern auftreten:

**Fehler 1. Art (Alpha-Fehler):**
- Nullhypothese wird abgelehnt, obwohl sie wahr ist (falsch positiv)

**Fehler 2. Art (Beta-Fehler):**
- Nullhypothese wird beibehalten, obwohl sie falsch ist (falsch negativ)

Ein Kompromiss muss oft zwischen der Minimierung dieser beiden Fehlerarten gefunden werden.

## Beispiel für z-Test

In diesem Beispiel wollen wir testen, ob der Mittelwert einer Stichprobe signifikant von einem angenommenen Mittelwert abweicht, und verwenden dabei den **z-Test**.

**Beispiel:**<br>
Wir haben eine Stichprobe von Testnoten und möchten testen, ob der Mittelwert der Stichprobe signifikant von einem angenommenen Mittelwert von 70 Punkten abweicht. Die Standardabweichung der Population beträgt 10 Punkte, und wir haben eine große Stichprobe.

<br>

Schritte des Hypothesentests:

1. **Nullhypothese** \( H_0 \): Der wahre Mittelwert der Population ist \( \mu_0 = 70 \).
2. **Alternativhypothese** \( H_1 \): Der wahre Mittelwert der Population unterscheidet sich von 70 (zweiseitiger Test).
3. **Signifikanzniveau** \( \alpha \): Wir verwenden \( \alpha = 0.05 \).
4. **z-Teststatistik**: Wird basierend auf den Daten berechnet.
5. **p-Wert**: Berechnung des p-Werts und Vergleich mit \( \alpha \).
6. **Entscheidung**: Wenn der p-Wert kleiner als \( \alpha \) ist, lehnen wir \( H_0 \) ab.

````python
import numpy as np
import scipy.stats as stats

# Gegebene Parameter
mu_0 = 70  # Hypothetischer Mittelwert (Nullhypothese)
sigma = 10  # Standardabweichung der Population (bekannt)
n = 50  # Stichprobengröße

# Simuliere Stichprobendaten (du kannst hier echte Daten verwenden)
np.random.seed(42)  # Setzt einen Zufallsstartwert für reproduzierbare Ergebnisse
stichprobe = np.random.normal(loc=72, scale=sigma, size=n)
mean_stichprobe = np.mean(stichprobe)

# Berechnung der z-Teststatistik
z = (mean_stichprobe - mu_0) / (sigma / np.sqrt(n))

# Berechnung des z-Werts und p-Werts (zweiseitiger Test)
p_wert = 2 * (1 - stats.norm.cdf(abs(z)))

# Ausgabe der Ergebnisse
print(f"Stichprobenmittelwert: {mean_stichprobe:.2f}")
print(f"z-Teststatistik: {z:.2f}")
print(f"p-Wert: {p_wert:.4f}")

# Entscheidung basierend auf einem Signifikanzniveau von 0.05
alpha = 0.05
if p_wert < alpha:
    print("Ergebnis: Die Nullhypothese wird abgelehnt.")
else:
    print("Ergebnis: Die Nullhypothese wird nicht abgelehnt.")
````


- wir simulieren eine Stichprobe von 50 Personen
   - deren Testnoten folgen einem normalverteilten Zufallsmuster
   - der Mittelwert der Stichprobe wird berechnet

- die z-Teststatistik wird berechnet, basierend auf
  - dem Mittelwert der Stichprobe,
  - dem hypothetischen Mittelwert (70),
  - der bekannten Standardabweichung der Population und
  - der Stichprobengröße


- wir berechnen den p-Wert für einen zweiseitigen Test
  - wenn p-Wert < Signifikanzniveau $( \alpha = 0.05 )$:
       - wir lehnen die Nullhypothese ab und sagen, dass der Mittelwert der Stichprobe signifikant von 70 abweicht
  - wenn p-Wert > 0.05:
     - Nullhypothese wird **nicht abgelehnt**
       - bedeutet, die Differenz zwischen dem Stichprobenmittelwert und dem angenommenen Mittelwert ist nicht signifikant



## Beispiel 1 für t-Test


In diesem Beispiel wollen wir testen, ob der Mittelwert einer Stichprobe signifikant von einem angenommenen Mittelwert (z.B. 70 Punkte) abweicht, wenn die Standardabweichung der Population nicht bekannt ist.

**Schritte des Hypothesentests:**

1. **Nullhypothese** \( H_0 \): Der wahre Mittelwert der Population ist \( \mu_0 = 70 \).
2. **Alternativhypothese** \( H_1 \): Der wahre Mittelwert der Population unterscheidet sich von 70 (zweiseitiger Test).
3. **Signifikanzniveau** \( \alpha \): Wir verwenden \( \alpha = 0.05 \).
4. **t-Teststatistik**: Wird basierend auf den Daten berechnet.
5. **p-Wert**: Berechnung des p-Werts und Vergleich mit \( \alpha \).
6. **Entscheidung**: Wenn der p-Wert kleiner als \( \alpha \) ist, lehnen wir \( H_0 \) ab.

````python
import numpy as np
import scipy.stats as stats

# Gegebene Parameter
mu_0 = 70  # Hypothetischer Mittelwert (Nullhypothese)
n = 15  # Stichprobengröße

# Simuliere Stichprobenwerte (du kannst hier echte Daten verwenden)
np.random.seed(42)  # Setzt einen Zufallsstartwert für reproduzierbare Ergebnisse
stichprobe = np.random.normal(loc=72, scale=12, size=n)
mean_stichprobe = np.mean(stichprobe)
std_stichprobe = np.std(stichprobe, ddof=1)  # Stichprobenstandardabweichung

# Berechnung der t-Teststatistik
t = (mean_stichprobe - mu_0) / (std_stichprobe / np.sqrt(n))

# Freiheitsgrade
df = n - 1

# Berechnung des p-Werts (zweiseitiger Test)
p_wert = 2 * (1 - stats.t.cdf(abs(t), df))

# Ausgabe der Ergebnisse
print(f"Stichprobenmittelwert: {mean_stichprobe:.2f}")
print(f"t-Teststatistik: {t:.2f}")
print(f"p-Wert: {p_wert:.4f}")

# Entscheidung basierend auf einem Signifikanzniveau von 0.05
alpha = 0.05
if p_wert < alpha:
    print("Ergebnis: Die Nullhypothese wird abgelehnt.")
else:
    print("Ergebnis: Die Nullhypothese wird nicht abgelehnt.")
````

- wir simulieren eine Stichprobe von 15 Personen
   - deren Testnoten folgen einem normalverteilten Zufallsmuster (Mittelwert 72 und Standardabweichung 12)
   
- die t-Teststatistik wird berechnet, basierend auf:
  - dem Mittelwert der Stichprobe,
  - dem hypothetischen Mittelwert $( \mu_0 = 70 \$,
  - der Stichprobenstandardabweichung und
  - der Stichprobengröße


- wir berechnen den p-Wert für einen zweiseitigen Test

  - wenn p-Wert < Signifikanzniveau $( \alpha = 0.05 )$:
    - wir lehnen die Nullhypothese ab und sagen, dass der Mittelwert der Stichprobe signifikant von 70 abweicht
  - wenn p-Wert > 0.05 ist
    - Nullhypothese wird **nicht abgelehnt**
      - bedeutet, dass die Differenz zwischen dem Stichprobenmittelwert und dem angenommenen Mittelwert nicht signifikant ist

**Hinweise:**
- Dieser t-Test wird als **zweiseitiger Test** durchgeführt, d.h., wir testen, ob der Mittelwert der Stichprobe signifikant **größer oder kleiner** als der hypothetische Mittelwert ist.
- Wenn man einen einseitigen Test durchführen möchte (z.B., um zu testen, ob der Mittelwert der Stichprobe **größer** als der hypothetische Mittelwert ist), müsste der p-Wert entsprechend angepasst werden:
  
  ```python
  # Einseitiger t-Test (Test auf größere Mittelwerte)
  p_wert = 1 - stats.t.cdf(t, df)
  ```
