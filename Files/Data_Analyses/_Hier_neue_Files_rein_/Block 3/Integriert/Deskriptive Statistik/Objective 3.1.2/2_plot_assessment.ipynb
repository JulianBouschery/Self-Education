# Boxplot
- zeigt die Verteilung eines Datensatzes
- gibt ersten Überblick über die Daten durch Darstellung von Zentralmaßen wie Spannweite, Median etc.
- hebt Ausreißer hervor, die die Statistik verzerren können


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Beispiel-Daten
np.random.seed(42)
data = {
    'Einkommen': np.random.normal(50000, 15000, 100)
}
df = pd.DataFrame(data)

# Boxplot
sns.boxplot(data=df['Einkommen'])
plt.title('Boxplot der Einkommen')
plt.show()
```

Der Boxplot zeigt die Verteilung des Einkommens

- blaue Kiste:
   - untere Kante: 1. Quartil
   - obere Kante: 3. Quartil
   - Höhe der Box: Interquartilsabstand (IQA)
   - durchgehende Linie in Box: Median
- Whiskers:
  - obere Antenne: größter Wert, der nicht zu den Ausreißern gehört
  - untere Antenne: kleinster Wert, der nicht zu den Ausreißern gehört
- Kringel:
  - Ausreißer

# Histogramm

- zeigt die Häufigkeitsverteilung von Daten

```python
# Histogramm
sns.histplot(df['Einkommen'], kde=True)
plt.title('Histogramm der Einkommen')
plt.show()
```

Das Histogramm zeigt die Häufigkeit der Einkommen in verschiedenen Bereichen. Die KDE-Linie gibt eine geglättete Verteilung der Daten an.

# Scatterplot

- zeigt die Beziehung zwischen zwei Variablen

Wir visualisieren den Zusammenhang zwischen Einkommen und Arbeitsstunden.

```python
# Beispiel-Daten erweitern
data['Arbeitsstunden'] = np.random.normal(40, 5, 100)
df = pd.DataFrame(data)

# Scatterplot
sns.scatterplot(x=df['Einkommen'], y=df['Arbeitsstunden'])
plt.title('Scatterplot: Einkommen vs. Arbeitsstunden')
plt.show()
```

Der Scatterplot zeigt keinen klaren linearen Zusammenhang zwischen Einkommen und Arbeitsstunden. Dies deutet darauf hin, dass andere Faktoren eine größere Rolle spielen könnten.

# Lineplot
- zeigt den Verlauf einer Variablen über die Zeit

Wir visualisieren die Änderung des Einkommens über 100 Tage.

```python
# Zeit hinzufügen
df['Tag'] = np.arange(1, 101)

# Lineplot
sns.lineplot(x=df['Tag'], y=df['Einkommen'])
plt.title('Lineplot: Einkommen über die Zeit')
plt.show()
```

Der Lineplot zeigt Schwankungen des Einkommens über die Zeit. Es gibt keine klaren Trends, was auf zufällige Variationen hinweisen könnte.

# Korrelation Heatmap
- zeigt die Stärke der Beziehung zwischen mehreren Variablen

Wir analysieren die Korrelation zwischen Einkommen, Arbeitsstunden und weiteren Variablen.

```python
# Korrelation berechnen
correlation_matrix = df.corr()

# Korrelation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korrelation Heatmap')
plt.show()
```

Die Heatmap zeigt die Stärke der linearen Beziehungen zwischen den Variablen. Werte nahe 1 oder -1 zeigen eine starke positive bzw. negative Korrelation an.
