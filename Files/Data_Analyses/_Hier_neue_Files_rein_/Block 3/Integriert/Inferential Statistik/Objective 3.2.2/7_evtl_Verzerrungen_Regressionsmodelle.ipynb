### Einschränkungen, Annahmen und potenzielle Verzerrungen in linearen und logistischen Regressionsmodellen

**1. Lineare Regression**

**Annahmen:**
- **Lineare Beziehung:** Es wird angenommen, dass zwischen den unabhängigen und abhängigen Variablen eine lineare Beziehung besteht. 
- **Homoskedastizität:** Die Varianz der Fehler ist konstant über alle Werte der unabhängigen Variablen.
- **Unabhängigkeit der Fehler:** Die Fehler (Residuals) sind unabhängig voneinander.
- **Normalverteilung der Fehler:** Die Fehler sollten normalverteilt sein.

**Einschränkungen:**
- **Nichtlineare Beziehungen:** Wenn die Beziehung zwischen Variablen nicht linear ist, kann das Modell ungenaue Vorhersagen liefern.
- **Multikollinearität:** Hohe Korrelationen zwischen unabhängigen Variablen können die Schätzungen der Koeffizienten verzerren.

**Beispiel:**
Stellen wir uns vor, wir versuchen, den Einfluss von Werbung auf den Umsatz zu modellieren. Wenn der Umsatz tatsächlich nur bei bestimmten Schwellenwerten der Werbung ansteigt, wäre ein lineares Modell möglicherweise nicht geeignet.

**Visualisierung:**
Ein Beispiel für eine nichtlineare Beziehung könnte ein Scatterplot sein, in dem die Punkte eine parabolische Kurve zeigen. Hier könnte eine lineare Regression fehlschlagen.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Daten erstellen
np.random.seed(0)
x = np.random.rand(100) * 10
y = 0.5 * x**2 + np.random.normal(size=x.size)

# Scatterplot
plt.scatter(x, y, color='blue', label='Datenpunkte')
plt.title('Scatterplot einer nichtlinearen Beziehung')
plt.xlabel('Werbung')
plt.ylabel('Umsatz')
plt.legend()
plt.grid()
plt.show()
````

**2. Logistische Regression**

**Annahmen:**
- **Binäre abhängige Variable:** Die abhängige Variable sollte binär sein (z.B. Ja/Nein).
- **Unabhängigkeit der Beobachtungen:** Jede Beobachtung sollte unabhängig von anderen sein.
- **Lineare Beziehung zwischen logit und unabhängigen Variablen:** Es wird angenommen, dass die logit-Transformation der abhängigen Variable linear in Bezug auf die unabhängigen Variablen ist.

**Einschränkungen:**
- **Multikollinearität:** Wie bei der linearen Regression kann hohe Korrelation zwischen unabhängigen Variablen die Schätzungen verzerren.
- **Überanpassung:** Wenn das Modell zu viele unabhängige Variablen enthält, kann es überangepasst werden und schlecht generalisieren.

**Beispiel:**
Angenommen, wir modellieren die Wahrscheinlichkeit, dass ein Kunde ein Produkt kauft, basierend auf Alter und Einkommen. Wenn wir nicht berücksichtigen, dass jüngere Kunden möglicherweise weniger kaufen, könnte das Modell diese Altersgruppe übersehen.

**Visualisierung:**
Ein Beispiel für die logistische Regression könnte eine Sigmoidkurve sein, die zeigt, wie die Wahrscheinlichkeit eines Kaufs mit dem Einkommen ansteigt.

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import seaborn as sns

# Daten erstellen
X, y = make_classification(n_samples=100, n_features=1, n_informative=1, n_redundant=0, random_state=0)
model = LogisticRegression()
model.fit(X, y)

# Vorhersagen
x_vals = np.linspace(X.min(), X.max(), 300)
y_probs = model.predict_proba(x_vals.reshape(-1, 1))[:, 1]

# Plot
plt.scatter(X, y, color='blue', label='Datenpunkte')
plt.plot(x_vals, y_probs, color='red', label='Logistische Regression')
plt.title('Logistische Regression')
plt.xlabel('Einkommen')
plt.ylabel('Wahrscheinlichkeit des Kaufs')
plt.legend()
plt.grid()
plt.show()
```

### Potenzielle Verzerrungen

**1. Auswahlverzerrung:** Wenn die Daten nicht repräsentativ für die gesamte Population sind, kann dies die Ergebnisse verzerren. Beispielsweise könnte eine Umfrage, die nur online durchgeführt wird, jüngere Nutzer überrepräsentieren.

**2. Messfehler:** Ungenaue Messungen der unabhängigen Variablen können die Schätzungen der Koeffizienten verfälschen.

### Auswirkungen auf die Ergebnisse

Die genannten Einschränkungen und Verzerrungen können die Vorhersagegenauigkeit und die Interpretierbarkeit der Modelle beeinträchtigen. Es ist wichtig, die Annahmen zu überprüfen und die Daten sorgfältig auszuwählen, um diese Probleme zu minimieren.
