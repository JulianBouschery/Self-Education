### Übung 1: Grundlegendes Liniendiagramm mit Matplotlib
Erstelle ein Liniendiagramm, das den Verlauf einer Funktion darstellt. Zum Beispiel kannst du den Verlauf der Sinusfunktion im Bereich von 0 bis 2π visualisieren.

**Anforderungen:**
- Verwende `numpy`, um die Daten für den Graphen zu generieren.
- Füge eine Beschriftung für die x- und y-Achse hinzu.
- Füge eine Legende und einen Titel zum Plot hinzu.

````python

import numpy as np
import matplotlib.pyplot as plt

# Daten generieren
x = np.linspace(start=0, stop=2*np.pi, num=100)  # 100 Werte für glatteren Graph
y = np.sin(x)  # Sinuswerte berechnen

# Liniendiagramm erstellen
plt.plot(x, y, label='sin(x)', color='blue')  # x und y für die plot Funktion angeben

# Achsen beschriften
plt.xlabel('x-Werte (0 bis 2π)')
plt.ylabel('y-Werte (sin(x))')

# Titel und Legende hinzufügen
plt.title('Sinusfunktion')
plt.legend()

# Diagramm anzeigen
plt.grid()  # Gitter hinzufügen für bessere Sichtbarkeit
plt.show()



````

### Übung 2: Balkendiagramm mit Matplotlib
Erstelle ein Balkendiagramm, das die Verkaufszahlen von fünf verschiedenen Produkten in einem Monat darstellt.

```python
produkte = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E']
verkaufszahlen = [150, 200, 130, 90, 240]
```

**Anforderungen:**
- Beschrifte die x- und y-Achse.
- Füge Farben und einen Titel hinzu.

````python
import numpy as np
import matplotlib.pyplot as plt

produkte = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E']
verkaufszahlen = [150, 200, 130, 90, 240]

plt.bar(produkte, verkaufszahlen, color='red')
plt.xlabel('Produkte')
plt.ylabel('Verkaufszahlen')
plt.title('Verkäufe der Produkte A bis E')
````

### Übung 3: Scatterplot mit Seaborn
Nutze einen echten Datensatz (z.B. den `tips`-Datensatz von Seaborn), um einen Scatterplot zu erstellen, der die Beziehung zwischen Rechnungsbetrag (`total_bill`) und Trinkgeld (`tip`) darstellt.

**Anforderungen:**
- Nutze den `seaborn.scatterplot()`-Befehl.
- Zeige zusätzlich die Kategorien der Tage als unterschiedliche Farben.
- Füge eine Trendlinie mit `sns.regplot()` hinzu.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = sns.load_dataset('tips')
print(data.columns)

sns.scatterplot(data, x='total_bill', y='tip', edgecolor='black', hue='day')

sns.regplot(data, x='total_bill', y='tip', scatter=False, color='gray')
````

### Übung 5: Heatmap mit Seaborn
Erstelle eine Heatmap, die eine Korrelationsmatrix eines Datensatzes (z.B. des `iris`-Datensatzes von Seaborn) visualisiert.

**Anforderungen:**
- Nutze die `sns.heatmap()`-Funktion.
- Füge eine Farbleiste (`colorbar`) hinzu.
- Stelle sicher, dass die Korrelationswerte im Diagramm angezeigt werden.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Iris-Datensatz laden
iris = sns.load_dataset('iris')

# Nur numerische Spalten auswählen
iris_numeric = iris.drop(columns=['species'])

# Korrelationsmatrix berechnen
correlation_matrix = iris_numeric.corr()

# Heatmap erstellen
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Titel hinzufügen
plt.title('Korrelationsmatrix der numerischen Iris-Daten')

# Plot anzeigen
plt.show()
````

### Übung 6: Mehrere Diagramme in einem Plot (Subplots)
Erstelle ein Diagramm mit mehreren Subplots, das die folgenden Diagrammtypen zeigt:
1. Liniendiagramm
2. Balkendiagramm
3. Histogramm

**Anforderungen:**
- Nutze `plt.subplot()` oder `plt.subplots()`, um mehrere Diagramme in einem zu erstellen.
- Füge Beschriftungen und Titel zu jedem der Subplots hinzu.


````python
import numpy as np
import matplotlib.pyplot as plt

# Daten für Liniendiagramm
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Daten für Balkendiagramm
produkte = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E']
verkaufszahlen = [150, 200, 130, 90, 240]

# Daten für Histogramm
data = np.random.randn(1000)

# Subplots erstellen
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Liniendiagramm
axs[0].plot(x, y)
axs[0].set_title('Liniendiagramm')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

# Balkendiagramm
axs[1].bar(produkte, verkaufszahlen, color='skyblue')
axs[1].set_title('Balkendiagramm')
axs[1].set_xlabel('Produkte')
axs[1].set_ylabel('Verkaufszahlen')

# Histogramm
axs[2].hist(data, bins=30, color='green')
axs[2].set_title('Histogramm')
axs[2].set_xlabel('Werte')
axs[2].set_ylabel('Häufigkeit')

# Platzierung der Subplots optimieren
plt.tight_layout()

# Plot anzeigen
plt.show()
````
