# Diskrete Daten
Diskrete Daten sind Daten, die nur bestimmte, meist zählbare Werte annehmen können. Sie entstehen oft durch Zählungen (z. B. Anzahl der Kinder in einer Familie) oder durch Kategorien (z. B. Geschlecht: männlich/weiblich).

## Beispiel
Stellen wir uns vor, du möchtest untersuchen, wie viele Stunden Menschen täglich am Computer arbeiten und ob sie an Übergewicht leiden (ja/nein). Hier ist „Übergewicht“ eine binäre, also diskrete, abhängige Variable.

Da die abhängige Variable diskret und binär ist, würden wir hier ein **logistisches Regressionsmodell** wählen. Bei der logistischen Regression versuchen wir vorherzusagen, wie wahrscheinlich es ist, dass jemand in die Kategorie „Übergewicht“ fällt, basierend auf der Anzahl der Computerstunden.

Die logistische Regression ist ideal, um Wahrscheinlichkeiten zu modellieren, da sie als Ergebnis Wahrscheinlichkeitswerte zwischen 0 und 1 liefert. Dies macht sie besonders nützlich bei Klassifizierungsproblemen mit diskreten Zielvariablen.

Man kann in einem Streudiagramm die Anzahl der Computerstunden auf der x-Achse und die Wahrscheinlichkeit für Übergewicht auf der y-Achse darstellen. Die logistische Kurve wird sich an die Datenpunkte anpassen und einen S-förmigen Verlauf zeigen, der die steigende oder fallende Wahrscheinlichkeit beschreibt.

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Beispielhafte Daten
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Stunden am Computer
y = np.array([0, 0, 0, 1, 1])  # Übergewicht: 0 = Nein, 1 = Ja

# Logistisches Regressionsmodell
model = LogisticRegression()
model.fit(X, y)
y_prob = model.predict_proba(X)[:, 1]

# Visualisierung
plt.scatter(X, y, color='blue', label='Datenpunkte')
plt.plot(X, y_prob, color='red', label='Logistische Kurve')
plt.xlabel('Stunden am Computer')
plt.ylabel('Wahrscheinlichkeit für Übergewicht')
plt.legend()
plt.show()
```

# Stetige Daten
Stetige Daten können jeden beliebigen Wert innerhalb eines Intervalls annehmen und sind oft das Ergebnis von Messungen (z. B. Körpergröße, Temperatur, Alter).

## Beispiel
Angenommen, du möchtest vorhersagen, wie das Einkommen eines Menschen von seiner Berufserfahrung abhängt. Hier ist „Einkommen“ eine stetige abhängige Variable.

In diesem Fall wäre ein **lineares Regressionsmodell** angebracht, da wir die Beziehung zwischen der stetigen abhängigen Variablen (Einkommen) und einer oder mehreren unabhängigen Variablen (z. B. Berufserfahrung) modellieren möchten. Das lineare Regressionsmodell versucht, die Datenpunkte durch eine lineare Funktion so gut wie möglich zu approximieren.

Ein einfaches Streudiagramm mit Berufserfahrung auf der x-Achse und Einkommen auf der y-Achse kann genutzt werden. Die lineare Regressionslinie, die durch die Datenpunkte verläuft, zeigt die Tendenz der Daten an.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Beispielhafte Daten
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Berufserfahrung
y = np.array([30000, 35000, 40000, 50000, 60000])  # Einkommen

# Lineares Regressionsmodell
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Visualisierung
plt.scatter(X, y, color='blue', label='Datenpunkte')
plt.plot(X, y_pred, color='red', label='Lineare Regression')
plt.xlabel('Berufserfahrung (Jahre)')
plt.ylabel('Einkommen (€)')
plt.legend()
plt.show()
```

---

# Wahl des Modells
- **Logistische Regression**: Wird gewählt, wenn die abhängige Variable diskret ist, insbesondere wenn es sich um Klassifikationsprobleme handelt (z. B. „ja/nein“, „gesund/krank“).
- **Lineare Regression**: Wird genutzt, wenn die abhängige Variable stetig ist, d. h. wenn wir eine kontinuierliche Beziehung modellieren möchten.
