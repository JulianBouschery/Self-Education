Die Anwendung von linearen und logistischen Regressionsmodellen in Python erfolgt in der Regel mithilfe von Bibliotheken wie 
- `scikit-learn`,
- `statsmodels` oder
- `numpy`

Wir sehen uns zwei einfache Beispiele für beide Regressionsmodelle zeigen, einschließlich der Modellanpassung und Parameterabschätzung. Die Ergebnisse veranschaulichen wir durch Plots.

# Was bedeutet Parameterabschätzung?

Mit Parameterabschätzung ist der Prozess gemeint, bei dem die Koeffizienten eines statistischen Modells auf Basis der gegebenen Daten geschätzt werden. In den Kontexten der linearen und logistischen Regression bezieht sich das auf die Schätzung der Parameter (auch als Koeffizienten oder Gewichte bezeichnet), die das Modell definieren.

## in der linearen Regression

In der linearen Regression ist das Modell durch die Gleichung
$
y = \beta_0 + \beta_1x + \epsilon
$
definiert. Hier sind die Parameter:
- **$\beta_0$** (Achsenabschnitt): Dieser Wert beschreibt, wo die Regressionsgerade die y-Achse schneidet. Es gibt den Wert von $y$ an, wenn $x = 0$ ist.
- **$\beta_1$** (Steigung): Dieser Wert beschreibt, wie stark sich $y$ ändert, wenn sich $x$ um eine Einheit ändert. 

Die Parameterabschätzung in der linearen Regression besteht darin, Werte für $\beta_0$ und $\beta_1$ zu finden, die die beste Anpassung an die beobachteten Daten liefern. Diese "beste Anpassung" wird in der Regel als Minimierung des mittleren quadratischen Fehlers zwischen den tatsächlichen und vorhergesagten Werten definiert. Methoden wie die **Kleinste-Quadrate-Methode** werden oft verwendet, um diese Parameter zu schätzen.

## in der logistischen Regression
Bei der logistischen Regression ist das Modell durch die Gleichung
$
p(x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x)}}
$
definiert, wobei $p(x)$ die Wahrscheinlichkeit angibt, dass das Ereignis eintritt (z. B. ob jemand Sport treibt).

Die Parameter, die geschätzt werden, sind wieder $\beta_0$ (Achsenabschnitt) und $\beta_1$ (Koeffizient), die bestimmen, wie stark der Prädiktor $x$ (z. B. Alter) die Wahrscheinlichkeit beeinflusst. 

Hier wird in der Regel die **Maximum-Likelihood-Methode** verwendet, um die Werte von $\beta_0$ und $\beta_1$ zu schätzen. Diese Methode maximiert die Wahrscheinlichkeit, dass die beobachteten Daten durch das Modell erklärt werden.

## Bedeutung der Parameterabschätzung
Die Parameterabschätzung ist der Kern der Modellanpassung, da sie es uns ermöglicht:
1. **Vorhersagen zu treffen**: Mit den geschätzten Parametern können wir für neue Werte von $x$ Vorhersagen für $y$ machen (bei der linearen Regression) oder Wahrscheinlichkeiten berechnen (bei der logistischen Regression).
2. **Modellinterpretation**: Die geschätzten Koeffizienten sagen uns, wie stark die unabhängigen Variablen (Prädiktoren) die abhängige Variable beeinflussen. Zum Beispiel sagt eine positive Steigung ($\beta_1$) in der linearen Regression, dass mit zunehmendem $x$ auch $y$ zunimmt.
3. **Hypothesentests**: In der Statistik können wir testen, ob die geschätzten Parameter signifikant von null verschieden sind, was darauf hinweist, dass eine Beziehung zwischen den Variablen besteht.

## Beispiel einer Parameterabschätzung in Python

Wenn wir auf das Beispiel der linearen Regression von vorhin zurückkommen:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Beispielhafte Daten: Körpergröße (in cm) und Gewicht (in kg)
X = np.array([[150], [160], [170], [180], [190]])  # Körpergröße
y = np.array([50, 60, 70, 80, 90])  # Gewicht

# Lineares Regressionsmodell erstellen und anpassen
model = LinearRegression()
model.fit(X, y)

# Parameterabschätzung
print(f'Intercept (β0): {model.intercept_}')  # Achsenabschnitt
print(f'Steigung (β1): {model.coef_[0]}')     # Steigung (Koeffizient)
```

Hier gibt die Methode `model.intercept_` den Achsenabschnitt $\beta_0$ und `model.coef_` die Steigung $\beta_1$ aus. Diese Werte repräsentieren die beste Schätzung der Parameter, die das lineare Modell beschreibt.

# Lineare Regression

Die lineare Regression wird verwendet, um eine Beziehung zwischen einer abhängigen (Ziel-) Variable und einer oder mehreren unabhängigen Variablen zu modellieren. Sie wird durch die Gleichung $ y = \beta_0 + \beta_1x + \epsilon $ beschrieben, wobei $ \beta_0 $ der Achsenabschnitt, $ \beta_1 $ der Koeffizient (Steigung) und $ \epsilon $ der Fehlerterm ist.

## Beispiel: Einfache lineare Regression
Nehmen wir an, wir haben ein Datenset mit der Körpergröße (in cm) und dem Gewicht (in kg) von Personen. Wir wollen das Gewicht als Funktion der Körpergröße vorhersagen.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Beispielhafte Daten: Körpergröße (in cm) und Gewicht (in kg)
X = np.array([[150], [160], [170], [180], [190]])  # Körpergröße
y = np.array([50, 60, 70, 80, 90])  # Gewicht

# Lineares Regressionsmodell erstellen und anpassen
model = LinearRegression()
model.fit(X, y)

# Vorhersagen treffen
y_pred = model.predict(X)

# Plot der Daten und der Regressionsgeraden
plt.scatter(X, y, color='blue', label='Datenpunkte')
plt.plot(X, y_pred, color='red', label='Regressionsgerade')
plt.xlabel('Körpergröße (cm)')
plt.ylabel('Gewicht (kg)')
plt.title('Lineare Regression: Vorhersage des Gewichts basierend auf der Körpergröße')
plt.legend()
plt.show()

# Parameterabschätzung
print(f'Intercept (β0): {model.intercept_}')
print(f'Steigung (β1): {model.coef_[0]}')
```

In diesem Beispiel passt das Modell eine gerade Linie an die Punkte an. Die Parameterabschätzung liefert den Achsenabschnitt und die Steigung der Linie, die die beste Anpassung an die Daten darstellt.

# Logistische Regression

Die logistische Regression wird häufig verwendet, um binäre Klassifikationsprobleme zu lösen. Anstelle einer kontinuierlichen Zielvariable, wie bei der linearen Regression, wird eine Wahrscheinlichkeit vorhergesagt, dass eine Beobachtung zu einer bestimmten Klasse gehört.

## Beispiel: Binäre Klassifikation
Angenommen, wir haben Daten über das Alter von Personen und ob sie regelmäßig Sport treiben (Ja = 1, Nein = 0). Wir möchten vorhersagen, ob jemand basierend auf dem Alter Sport treibt.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Beispielhafte Daten: Alter und ob jemand Sport treibt (1 = Ja, 0 = Nein)
X = np.array([[18], [25], [32], [40], [50], [60]])  # Alter
y = np.array([0, 0, 1, 1, 1, 0])  # 1 = treibt Sport, 0 = treibt keinen Sport

# Logistisches Regressionsmodell erstellen und anpassen
log_model = LogisticRegression()
log_model.fit(X, y)

# Erstelle einen feineren Bereich für die x-Werte zur Interpolation
X_new = np.linspace(15, 65, 100).reshape(-1, 1)  # Mehr Datenpunkte
y_prob_new = log_model.predict_proba(X_new)[:, 1]  # Wahrscheinlichkeiten für die neuen Punkte

# Plot der Daten und der Wahrscheinlichkeitskurve
plt.scatter(X, y, color='blue', label='Datenpunkte')
plt.plot(X_new, y_prob_new, color='red', label='Vorhergesagte Wahrscheinlichkeit (Sigmoid-Kurve)', linewidth=2)
plt.axhline(0.5, color='green', linestyle='--', label='Entscheidungsgrenze (0.5)')
plt.xlabel('Alter')
plt.ylabel('Wahrscheinlichkeit, Sport zu treiben')
plt.title('Logistische Regression: Sporttreiben basierend auf dem Alter')
plt.ylim(-0.1, 1.1)  # Y-Achse anpassen, um die Sigmoidform zu betonen
plt.legend()
plt.show()

# Parameterabschätzung
print(f'Intercept (β0): {log_model.intercept_[0]}')
print(f'Koeffizient (β1): {log_model.coef_[0][0]}')
```

In diesem Beispiel gibt das logistische Modell die Wahrscheinlichkeit aus, dass eine Person Sport treibt, basierend auf ihrem Alter. Die Wahrscheinlichkeitskurve folgt der logistischen Funktion und zeigt eine S-Kurvenform.

# Zusammenfassung der Schritte
1. **Daten vorbereiten**: Unabhängige und abhängige Variablen in geeigneten Arrays.
2. **Modell anpassen**: Mit der `fit`-Methode das Modell an die Daten anpassen.
3. **Vorhersagen treffen**: Mithilfe der `predict`-Methode oder `predict_proba` für logistische Regression.
4. **Parameter abschätzen**: Auf die geschätzten Modellparameter wie `intercept_` und `coef_` zugreifen.
5. **Visualisierung**: Die Daten und die Vorhersagen/Modelle visualisieren.

Beide Modelle sind sehr nützlich, um Vorhersagen zu treffen, wobei die lineare Regression eher für kontinuierliche Daten und die logistische Regression für binäre (oder kategoriale) Klassifikationsaufgaben verwendet wird.
