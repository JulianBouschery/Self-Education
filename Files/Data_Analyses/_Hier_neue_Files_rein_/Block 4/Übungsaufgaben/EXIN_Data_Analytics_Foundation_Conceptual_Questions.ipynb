# EXIN (Data Analytics Foundation) - Conceptual Questions

## 1. Daten in Erkenntnisse umwandeln (10%)

### 1.1 Begriffe und Prozess (7.5%)

1. **Was versteht man unter Datenanalyse? Beschreiben Sie den allgemeinen Prozess der Datenanalyse von der Datenerhebung bis zur Entscheidungsfindung.**

2. **Wie wird Datenanalyse in verschiedenen Branchen genutzt? Nennen Sie Beispiele für die praktische Anwendung von Datenanalyse in unterschiedlichen Bereichen.**

### 1.2 Geschäftsanalytik (BI) (2.5%)

3. **Was versteht man unter Business Intelligence (BI)? Wie hilft BI Unternehmen dabei, datengetriebene Entscheidungen zu treffen?**

4. **Nennen Sie gängige Business Intelligence-Tools und deren Hauptanwendungsbereiche.**

---

## 2. Sammeln, organisieren, verwalten (32.5%)

### 2.1 Datensammlung (10%)

5. **Welche Methoden der Datensammlung gibt es? Beschreiben Sie, wie Unternehmen Daten über verschiedene Kanäle sammeln.**

6. **Was sind die ethischen Überlegungen bei der Datensammlung? Welche Vorschriften und Richtlinien sind zu beachten?**

### 2.2 Daten speichern (12.5%)

7. **Was versteht man unter relationalen und nicht-relationalen Datenbanken? Welche Unterschiede gibt es in der Art und Weise, wie Daten gespeichert werden?**

8. **Erklären Sie den Begriff der Datenarchivierung. Welche Techniken werden zur Speicherung großer Datenmengen verwendet?**

### 2.3 Variablen (10%)

9. **Was ist der Unterschied zwischen einer unabhängigen und einer abhängigen Variable? Geben Sie Beispiele aus der Praxis.**

10. **Was versteht man unter kategorischen und kontinuierlichen Variablen? Wie beeinflusst dies die Auswahl von Analysemethoden?**

---

## 3. Bereinigen (7.5%)

### 3.1 Daten-Scrubbing (7.5%)

11. **Was versteht man unter Datenbereinigung? Beschreiben Sie den Prozess des "Data Scrubbing".**

12. **Warum ist Datenbereinigung wichtig für eine genaue Analyse? Welche Techniken gibt es, um Daten zu bereinigen?**

---

## 4. Analysieren (35%)

### 4.1 Statistik (5%)

13. **Erklären Sie die Begriffe "Durchschnitt", "Median" und "Standardabweichung". Wann wird welcher Wert zur Datenanalyse verwendet?**

14. **Wie berechnen Sie die Varianz eines Datensatzes? Warum ist die Varianz ein wichtiger Indikator in der Statistik?**

### 4.2 Data Mining (2.5%)

15. **Was versteht man unter Data Mining? Wie wird Data Mining eingesetzt, um Muster in großen Datensätzen zu finden?**

16. **Beschreiben Sie die Schritte eines typischen Data-Mining-Prozesses.**

### 4.3 Maschinelles Lernen (10%)

17. **Was ist der Unterschied zwischen überwachten und unüberwachten Lernverfahren im maschinellen Lernen?**

18. **Nennen Sie einige Algorithmen des überwachten Lernens und erklären Sie, wann diese eingesetzt werden.**

### 4.4 Algorithmen (17.5%)

19. **Was sind Algorithmen in der Datenanalyse? Geben Sie Beispiele für gängige Algorithmen, die in der Datenanalyse verwendet werden.**

20. **Erklären Sie den Unterschied zwischen einem Entscheidungsbaum und einer linearen Regression. Wann würde man jeweils welchen Algorithmus einsetzen?**

---

## 5. Visualisieren (15%)

### 5.1 Explanativ und explorativ (2.5%)

21. **Was ist der Unterschied zwischen explanativer und explorativer Datenvisualisierung? Geben Sie Beispiele, wann jede Methode verwendet wird.**

### 5.2 Diagramme und Plots (7.5%)

22. **Welche Arten von Diagrammen und Plots gibt es in der Datenvisualisierung? Beschreiben Sie die häufigsten Darstellungsformen und deren Verwendung.**

23. **Wie erstellt man ein Streudiagramm und ein Balkendiagramm in Python?**

### 5.3 Gestaltung und Werkzeuge (5%)

24. **Welche Werkzeuge werden häufig für die Datenvisualisierung verwendet? Nennen Sie Vor- und Nachteile von Tools wie Tableau, Power BI und Matplotlib.**

25. **Was sind wichtige Designprinzipien bei der Erstellung von Datenvisualisierungen? Erklären Sie die Rolle von Farbe, Skalierung und Layout in der Datenpräsentation.**
