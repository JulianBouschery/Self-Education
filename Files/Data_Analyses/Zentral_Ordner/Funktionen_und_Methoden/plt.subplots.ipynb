# plt.subplots()

`plt.subplots()` ist eine Funktion in **Matplotlib**, die eine Figur (ein gesamtes Diagramm) und ein **Raster von Unterdiagrammen (Subplots)** erstellt, die dann mit verschiedenen Plots gefüllt werden können. Diese Funktion ist besonders nützlich, wenn man mehrere Plots auf einmal anzeigen möchte, z. B. in einem Gitternetz von Diagrammen.

### Grundsätzliche Syntax

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6), dpi=100)
```

### Parameter von `plt.subplots()`

- **`nrows`** und **`ncols`**: Anzahl der Zeilen und Spalten für das Raster der Subplots.
  - Beispiel: `nrows=2` und `ncols=3` erzeugt ein Raster mit 2 Zeilen und 3 Spalten (also insgesamt 6 Subplots).
  - Wenn man nur einen Plot benötigt, kann man `nrows=1` und `ncols=1` verwenden.

- **`figsize`**: Ein Tupel, das die **Größe der gesamten Figur** in Zoll angibt, z. B. `figsize=(15, 7)`.
  - Dies beeinflusst die Größe aller Subplots zusammen.

- **`dpi`**: Die **Auflösung** der Figur in "dots per inch". Höhere DPI-Werte führen zu detaillierteren, aber größeren Grafiken.

- **`sharex`** und **`sharey`**: Bestimmen, ob Subplots die gleiche X- oder Y-Achse teilen sollen.
  - `sharex=True` bedeutet, dass alle Subplots dieselbe X-Achse verwenden. Dies ist nützlich, wenn man denselben Bereich auf mehreren Subplots zeigen will.

### Rückgabewerte

- **`fig`**: Das **Figure-Objekt** repräsentiert das gesamte Diagramm, das alle Subplots enthält. Es ist wie eine leere Leinwand.
- **`ax`**: Ein **Array von Axes-Objekten**, die die einzelnen Subplots repräsentieren. 
  - Wenn `nrows` und `ncols` beide 1 sind, ist `ax` ein einzelnes `Axes`-Objekt.
  - Bei mehreren Zeilen und/oder Spalten wird `ax` zu einem 2D-Array von `Axes`-Objekten.

### Beispiel: 1 Zeile, 2 Spalten

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot([1, 2, 3], [1, 4, 9])
ax[1].plot([1, 2, 3], [1, 2, 3])

plt.show()
````

- **Beschreibung**: Erstellt eine Figur mit einer Zeile und zwei Spalten.
- `ax[0]` und `ax[1]` sind die beiden einzelnen Subplots.

### Beispiel: 2 Zeilen, 3 Spalten

````python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
ax[0, 0].plot([1, 2, 3], [1, 4, 9])
ax[1, 2].plot([1, 2, 3], [1, 2, 3])

plt.tight_layout()
plt.show()
````

- **Beschreibung**: Erstellt ein 2x3-Gitter.
- Der Zugriff auf Subplots erfolgt über `ax[Zeile, Spalte]` (z. B. `ax[0, 0]` für das erste Diagramm in der ersten Zeile).

### Zusammenfassung

- **`plt.subplots()`** ist ideal für die Erstellung mehrerer Subplots in einem Rasterformat.
- Man kann die Anzahl der Zeilen und Spalten sowie die Größe des gesamten Plots festlegen.
- `fig` steuert das Gesamtlayout, während `ax` einzelne Plots enthält, auf die man direkt zugreifen und sie anpassen kann.
