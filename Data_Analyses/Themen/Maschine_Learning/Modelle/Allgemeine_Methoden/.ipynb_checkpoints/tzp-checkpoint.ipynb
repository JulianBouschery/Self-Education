# t- vs. z-Statistik und p-Wert

## z-Statistik

## t-Statistik

Die **t-Statistik** ist eine Teststatistik, die in der Inferenzstatistik verwendet wird, um zu beurteilen, ob es einen signifikanten Unterschied zwischen dem Mittelwert einer Stichprobe und einem bekannten Wert (z.B. dem Mittelwert der Grundgesamtheit) gibt, oder um den Unterschied zwischen den Mittelwerten von zwei Stichproben zu überprüfen. Sie wird vor allem im Zusammenhang mit dem **t-Test** verwendet.

Die t-Statistik misst, wie weit der Mittelwert der Stichprobe von dem Mittelwert der Grundgesamtheit entfernt ist, **in Einheiten der Standardfehler**. Mit anderen Worten, sie gibt an, wie viele Standardabweichungen der Mittelwert der Stichprobe vom erwarteten Mittelwert (laut Nullhypothese) entfernt ist.

## p-Wert

Der **p-Wert** ist ein zentrales Konzept in der Inferenzstatistik und gibt an, wie wahrscheinlich es ist, die beobachteten Daten (oder extremere Werte) zu erhalten, **unter der Annahme, dass die Nullhypothese wahr ist**. Er hilft dabei, zu entscheiden, ob die Nullhypothese (keine Wirkung oder kein Unterschied) abgelehnt werden sollte oder nicht.

Wenn du einen t-Test durchführst, bei dem die Nullhypothese lautet, dass der Mittelwert einer Stichprobe gleich dem Mittelwert der Grundgesamtheit ist, dann zeigt der p-Wert, wie wahrscheinlich die Unterschiede zwischen den beiden Mittelwerten rein zufällig sind.

## Zusammenfassung:

**t-Statistik**: Sagt uns, wie viele Standardfehler der Mittelwert der Stichprobe vom Mittelwert der Grundgesamtheit entfernt ist.

**p-Wert**: Zeigt an, ob dieser Unterschied signifikant ist oder nicht.

## Fall 1: Ein niedriger p-Wert

Eine **hohe t-Statistik** bedeutet, dass der Mittelwert der Stichprobe deutlich vom Mittelwert der Grundgesamtheit abweicht. Das deutet darauf hin, dass der Unterschied wahrscheinlich nicht zufällig ist.

Ein **niedriger p-Wert** (typischerweise < 0,05) deutet darauf hin, dass die beobachteten Daten unwahrscheinlich wären, wenn die Nullhypothese korrekt ist. Das bedeutet, es gibt genügend Beweise, um die Nullhypothese abzulehnen.

Je höher die t-Statistik desto niedriger der p-Wert.???

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Beispieldaten: Stichprobe
stichprobe = [22, 24, 25, 23, 26, 27, 28, 30, 22, 25]

# Grundgesamtheit: Mittelwert 24
popmean = 25

# t-Test durchführen
t_stat, p_value = stats.ttest_1samp(stichprobe, popmean)
print(f't-Statistik: {t_stat}, p-Wert: {p_value}')

# Entscheidung basierend auf dem p-Wert
alpha = 0.05
if p_value < alpha:
    print('Die Nullhypothese wird abgelehnt (es gibt einen signifikanten Unterschied).')
else:
    print('Die Nullhypothese wird beibehalten (kein signifikanter Unterschied).')

# Visualisierung des t-Tests
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=len(stichprobe) - 1)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='t-Verteilung', color='blue')

# Kritische Werte für die t-Verteilung
t_critical = stats.t.ppf(1 - alpha / 2, df=len(stichprobe) - 1)

# Markiere die kritischen Bereiche
plt.fill_between(x, 0, y, where=(x <= -t_critical) | (x >= t_critical), color='red', alpha=0.5, label='Ablehnungsbereich')

# Markiere die t-Statistik
plt.axvline(t_stat, color='green', linestyle='--', label=f't-Statistik = {t_stat:.2f}', linewidth=2)

# Titel und Achsenbeschriftungen
plt.title('Visualisierung des t-Tests', fontsize=14)
plt.xlabel('t-Wert')
plt.ylabel('Dichte')

# Legende
plt.legend()

# Ergebnisanzeige
plt.show()
```

Wenn **p-Wert < 0.05**: Es gibt weniger als 5 % Wahrscheinlichkeit, dass die beobachteten Daten zufällig unter der Nullhypothese zustande gekommen sind. → **Die Nullhypothese wird abgelehnt**.

### Fall 2: Ein hoher p-Wert

Eine **niedrige t-Statistik** bedeutet, dass der Mittelwert der Stichprobe nahe am Mittelwert der Grundgesamtheit liegt, was darauf hindeutet, dass es keinen signifikanten Unterschied gibt.

Ein **hoher p-Wert** (typischerweise > 0,05) deutet darauf hin, dass die beobachteten Daten mit der Nullhypothese vereinbar sind. In diesem Fall gibt es nicht genügend Beweise, um die Nullhypothese abzulehnen.

Je niedriger die t-Statistik desto höher der p-Wert.???

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Beispieldaten: Stichprobe
stichprobe = [22, 24, 25, 23, 26, 27, 28, 30, 22, 25]

# Grundgesamtheit: Mittelwert 24
popmean = 24

# t-Test durchführen
t_stat, p_value = stats.ttest_1samp(stichprobe, popmean)
print(f't-Statistik: {t_stat}, p-Wert: {p_value}')

# Entscheidung basierend auf dem p-Wert
alpha = 0.05
if p_value < alpha:
    print('Die Nullhypothese wird abgelehnt (es gibt einen signifikanten Unterschied).')
else:
    print('Die Nullhypothese wird beibehalten (kein signifikanter Unterschied).')

# Visualisierung des t-Tests
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=len(stichprobe) - 1)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='t-Verteilung', color='blue')

# Kritische Werte für die t-Verteilung
t_critical = stats.t.ppf(1 - alpha / 2, df=len(stichprobe) - 1)

# Markiere die kritischen Bereiche
plt.fill_between(x, 0, y, where=(x <= -t_critical) | (x >= t_critical), color='red', alpha=0.5, label='Ablehnungsbereich')

# Markiere die t-Statistik
plt.axvline(t_stat, color='green', linestyle='--', label=f't-Statistik = {t_stat:.2f}', linewidth=2)

# Titel und Achsenbeschriftungen
plt.title('Visualisierung des t-Tests', fontsize=14)
plt.xlabel('t-Wert')
plt.ylabel('Dichte')

# Legende
plt.legend()

# Ergebnisanzeige
plt.show()
```

Wenn **p-Wert > 0.05**: Es gibt mehr als 5 % Wahrscheinlichkeit, dass die beobachteten Daten durch Zufall entstanden sind. → **Die Nullhypothese wird beibehalten**.

### Formel für die t-Statistik:

Für einen **Ein-Stichproben-t-Test** lautet die Formel:

$$
t = \frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}
$$

- $\bar{X}$ : Der Mittelwert der Stichprobe
- $\mu$ : Der Mittelwert der Grundgesamtheit (laut Nullhypothese)
- $s$ : Die Standardabweichung der Stichprobe
- $n$ : Die Anzahl der Stichprobenwerte (Stichprobengröße)

### Rechnung in Schritten:

```python
# Gegebene Werte
stichprobe = [22, 24, 25, 23, 26, 27, 28, 30, 22, 25]
popmean = 24
```

```python
# Berechnung des Mittelwerts
mean1 = sum(stichprobe)/len(stichprobe)
print(mean1)
mean2 = np.array(stichprobe).mean()
print(mean2)
```

```python
# Berechnung der Differenz der beiden Mittelwerte
diff = mean2 - popmean
print(diff)
```

```python
# Berechnung der Standardabweichung der Stichprobe
std_ = np.std(stichprobe, ddof=1) # Macht was genau?
print(std_)
```

```python
# Berechnung der Wurzel aus n
import math

n = len(stichprobe)
print(n)

squaerroot = math.sqrt(n)
print(squaerroot)
```

```python
# Berechnung des Standardfehlers 
std_error = std_ / squaerroot
print(std_error)
```

```python
# Berechnung der t-Statistik
t = diff / std_error
print(t)
```

### Zusammenfassung

```python
# Gegebene Werte
stichprobe = [22, 24, 25, 23, 26, 27, 28, 30, 22, 25]
popmean = 24

# Berechnung des Mittelwerts
mean = np.array(stichprobe).mean()

# Berechnung der Differenz der beiden Mittelwerte
diff = mean - popmean

# Berechnung der Standardabweichung der Stichprobe
std_ = np.std(stichprobe, ddof=1) # Macht was genau? 

# Berechnung der Wurzel aus n
import math
n = len(stichprobe)
squaerroot = math.sqrt(n)

# Berechnung des Standardfehlers 
std_error = std_ / squaerroot

# Berechnung der t-Statistik
t = diff / std_error
print(t)
```
