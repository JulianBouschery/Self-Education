# Speedtest

```python
# Importiere benötigte Bibliotheken
import numpy as np
import pandas as pd
import time

# Python-Listen vs. NumPy Arrays vs Pandas Series
list_a = [1, 2, 3]
array_a = np.array([1, 2, 3])
series_a = pd.Series([1, 2, 3])

# Performance-Vergleich
print("\nPerformance-Vergleich:")

start_time = time.time()
sum([i for i in range(1000000)])  # Listen
print("Listen Summe Zeit:", time.time() - start_time)

start_time = time.time()
np.sum(np.arange(1000000))  # NumPy Arrays
print("NumPy Summe Zeit:", time.time() - start_time)

start_time = time.time()
pd.Series(np.arange(1000000)).sum()  # Pandas Series
print("Pandas Series Summe Zeit:", time.time() - start_time)

# Unterschied zwischen NumPy-Arrays und Pandas Series
print("\nStrukturelle Unterschiede:")
print("NumPy Array:", array_a)
print("Pandas Series:\n", series_a)

# Zugriff auf einzelne Elemente
print("\nZugriff auf Elemente:")
print("Element in NumPy Array:", array_a[0])  # Zugriff mit Index
print("Element in Pandas Series mit Index:", series_a[0])  # Zugriff mit Index
print("Element in Pandas Series mit Label:", series_a.loc[0])  # Zugriff mit Label

# Struktur der Series (mit Index)
print("\nPandas Series Index:")
print(series_a.index)  # Zeigt den Index an

# Unterschied zwischen NumPy-Arrays und Pandas DataFrames
print("\nUnterschied zwischen NumPy-Arrays und Pandas DataFrames:")
data = {
    'Spalte1': [1, 2, 3],
    'Spalte2': [4, 5, 6]
}
df = pd.DataFrame(data)
print("Pandas DataFrame:\n", df)

# Unterschiede in den Datentypen und Flexibilität
print("\nUnterschiede in Datentypen:")
print("NumPy Array dtype:", array_a.dtype)
print("Pandas Series dtype:", series_a.dtype)

# Flexibilität von Pandas Series
series_b = pd.Series([1, 2, 3], index=['a', 'b', 'c'])  # Custom Index
print("\nPandas Series mit benutzerdefiniertem Index:\n", series_b)
```
